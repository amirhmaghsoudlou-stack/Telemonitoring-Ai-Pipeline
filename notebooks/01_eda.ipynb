```python
import numpy as np
import pandas as pd

# مرتب‌سازی زمانی
d = df_clean.sort_values("RelativeTimeMilliseconds").copy()

# پارامترها
THRESH_SPO2 = 98          # آستانه افت
HORIZON_SEC = 5 * 60      # افق پیش‌بینی = 5 دقیقه

# نرخ نمونه‌برداری را از روی داده تخمین می‌زنیم (برای اینکه به 1Hz وابسته نباشیم)
dt_ms = d["RelativeTimeMilliseconds"].diff().median()
dt_ms
```

```python
# تعداد گام‌ها در افق 5 دقیقه
h = int((HORIZON_SEC * 1000) / dt_ms)
```

```python
# فقط روی SpO2 معتبر کار کنیم
d["SpO2"] = pd.to_numeric(d["SpO2"], errors="coerce")

# مینیمم SpO2 در بازه [اکنون تا 5 دقیقه آینده] (تقریبی بر اساس گام‌ها)
future_min_spo2 = d["SpO2"][::-1].rolling(window=h, min_periods=1).min()[::-1]

# Label: اگر در افق 5 دقیقه به زیر آستانه برسد = 1
d["y_spo2_drop"] = (future_min_spo2 < THRESH_SPO2).astype(int)

d[["t_min","SpO2","y_spo2_drop"]].head(10)
```

```python
d["y_spo2_drop"].value_counts(normalize=True)
```

```python
WIN_SEC = 60
w = int((WIN_SEC * 1000) / dt_ms)

# تبدیل عددی
d["HR"] = pd.to_numeric(d["HR"], errors="coerce")
d["SpO2"] = pd.to_numeric(d["SpO2"], errors="coerce")

# ویژگی‌های rolling برای HR
d["HR_mean"] = d["HR"].rolling(w, min_periods=w).mean()
d["HR_std"]  = d["HR"].rolling(w, min_periods=w).std()
d["HR_min"]  = d["HR"].rolling(w, min_periods=w).min()
d["HR_max"]  = d["HR"].rolling(w, min_periods=w).max()

# ویژگی‌های rolling برای SpO2
d["SpO2_mean"] = d["SpO2"].rolling(w, min_periods=w).mean()
d["SpO2_std"]  = d["SpO2"].rolling(w, min_periods=w).std()
d["SpO2_min"]  = d["SpO2"].rolling(w, min_periods=w).min()
d["SpO2_max"]  = d["SpO2"].rolling(w, min_periods=w).max()

# شیب ساده در 60 ثانیه (Δvalue / Δtime)
d["HR_slope"] = (d["HR"] - d["HR"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))
d["SpO2_slope"] = (d["SpO2"] - d["SpO2"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))

feature_cols = ["HR_mean","HR_std","HR_min","HR_max","HR_slope",
                "SpO2_mean","SpO2_std","SpO2_min","SpO2_max","SpO2_slope"]

# دیتاست مدل (ردیف‌هایی که feature کامل ندارند حذف)
model_df = d.dropna(subset=feature_cols + ["y_spo2_drop"]).copy()

model_df[feature_cols + ["y_spo2_drop"]].head()
```

```python
from sklearn.model_selection import train_test_split

X = model_df[feature_cols].copy()
y = model_df["y_spo2_drop"].copy()

# تقسیم زمانی: بدون shuffle
split_idx = int(len(model_df) * 0.8)

X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]

X_train.shape, X_test.shape
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

clf = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("lr", LogisticRegression(max_iter=2000, class_weight="balanced", random_state=42))
])

clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

y_pred = clf.predict(X_test)
y_prob = clf.predict_proba(X_test)[:, 1]

print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
print("\nReport:\n", classification_report(y_test, y_pred, digits=3))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))
```

```python
import numpy as np
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    confusion_matrix, classification_report, roc_auc_score,
    precision_recall_curve, precision_score, recall_score, f1_score
)

BASE_PATH = r"E:\projects\Telemonitoring-Ai-Pipeline\data\uq-vitalsigns\uqvitalsignsdata"

THRESH_SPO2 = 98          # آستانه افت اکسیژن
HORIZON_SEC = 5 * 60      # افق پیش‌بینی (۵ دقیقه)
WIN_SEC = 60              # طول پنجره ویژگی‌ها (۶۰ ثانیه)

def load_case(case_id: int):
    cid = f"case{case_id:02d}"
    path = fr"{BASE_PATH}\{cid}\uq_vsd_{cid}_trenddata.csv"
    df = pd.read_csv(path, sep=",", header=0, engine="python", on_bad_lines="skip", index_col=False)
    return df, path

def build_dataset(df):
    d = df.copy()
    # تبدیل به عدد
    d["RelativeTimeMilliseconds"] = pd.to_numeric(d["RelativeTimeMilliseconds"], errors="coerce")
    d["HR"] = pd.to_numeric(d["HR"], errors="coerce")
    d["SpO2"] = pd.to_numeric(d["SpO2"], errors="coerce")
    d = d.dropna(subset=["RelativeTimeMilliseconds"]).sort_values("RelativeTimeMilliseconds")

    # تخمین گام زمانی
    dt_ms = d["RelativeTimeMilliseconds"].diff().median()
    if pd.isna(dt_ms) or dt_ms <= 0:
        return None

    # ساخت برچسب: آیا در ۵ دقیقه آینده SpO2 < 98 می‌شود؟
    h = int((HORIZON_SEC * 1000) / dt_ms)
    future_min_spo2 = d["SpO2"][::-1].rolling(window=h, min_periods=1).min()[::-1]
    d["y"] = (future_min_spo2 < THRESH_SPO2).astype(int)

    # پنجره ویژگی‌ها (۶۰ ثانیه)
    w = int((WIN_SEC * 1000) / dt_ms)

    d["HR_mean"] = d["HR"].rolling(w, min_periods=w).mean()
    d["HR_std"]  = d["HR"].rolling(w, min_periods=w).std()
    d["HR_min"]  = d["HR"].rolling(w, min_periods=w).min()
    d["HR_max"]  = d["HR"].rolling(w, min_periods=w).max()

    d["SpO2_mean"] = d["SpO2"].rolling(w, min_periods=w).mean()
    d["SpO2_std"]  = d["SpO2"].rolling(w, min_periods=w).std()
    d["SpO2_min"]  = d["SpO2"].rolling(w, min_periods=w).min()
    d["SpO2_max"]  = d["SpO2"].rolling(w, min_periods=w).max()

    d["HR_slope"] = (d["HR"] - d["HR"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))
    d["SpO2_slope"] = (d["SpO2"] - d["SpO2"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))

    feature_cols = ["HR_mean","HR_std","HR_min","HR_max","HR_slope",
                    "SpO2_mean","SpO2_std","SpO2_min","SpO2_max","SpO2_slope"]

    model_df = d.dropna(subset=feature_cols + ["y"]).copy()
    return model_df, feature_cols

def pick_threshold_on_train(y_true, y_prob, target_recall=0.90):
    prec, rec, thr = precision_recall_curve(y_true, y_prob)
    pr_df = pd.DataFrame({"threshold": np.r_[thr, np.nan], "precision": prec, "recall": rec})
    c = pr_df.dropna(subset=["threshold"]).copy()
    c = c[c["recall"] >= target_recall]
    if len(c) == 0:
        return 0.5  # اگر نشد، برگرد به پیش‌فرض
    # بهترین دقت (و سپس بهترین فراخوانی)
    best_row = c.sort_values(["precision", "recall"], ascending=False).iloc[0]
    return float(best_row["threshold"])

def eval_with_threshold(y_true, y_prob, t):
    y_pred = (y_prob >= t).astype(int)
    cm = confusion_matrix(y_true, y_pred)
    p1 = precision_score(y_true, y_pred, zero_division=0)
    r1 = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    # استخراج TN, FP, FN, TP
    tn, fp, fn, tp = cm.ravel()
    return {"threshold": t, "precision_1": p1, "recall_1": r1, "f1_1": f1,
            "TN": tn, "FP": fp, "FN": fn, "TP": tp}

results = []

CASE_LIST = [1, 2, 3, 4, 5]   # فعلاً ۵ کیس اول

for case_id in CASE_LIST:
    df, path = load_case(case_id)
    built = build_dataset(df)
    if built is None:
        results.append({"case": case_id, "status": "failed_build"})
        continue

    model_df, feature_cols = built

    # نسبت داده گمشده برای HR و SpO2 (صرفاً گزارشی)
    miss_hr = pd.to_numeric(df["HR"], errors="coerce").isna().mean()
    miss_spo2 = pd.to_numeric(df["SpO2"], errors="coerce").isna().mean()

    # نسبت کلاس‌ها
    class_ratio_1 = model_df["y"].mean()

    # تقسیم زمانی ۸۰/۲۰
    split_idx = int(len(model_df) * 0.8)
    train_df = model_df.iloc[:split_idx].copy()
    test_df  = model_df.iloc[split_idx:].copy()

    X_train = train_df[feature_cols]
    y_train = train_df["y"]
    X_test  = test_df[feature_cols]
    y_test  = test_df["y"]

    # مدل پایه
    clf = Pipeline(steps=[
        ("scaler", StandardScaler()),
        ("lr", LogisticRegression(max_iter=2000, class_weight="balanced", random_state=42))
    ])
    clf.fit(X_train, y_train)

    # احتمال‌ها
    prob_train = clf.predict_proba(X_train)[:, 1]
    prob_test  = clf.predict_proba(X_test)[:, 1]

    # آستانه هشدار را فقط با آموزش انتخاب می‌کنیم
    tuned_t = pick_threshold_on_train(y_train, prob_train, target_recall=0.90)

    # ارزیابی با آستانه پیش‌فرض و آستانه تنظیم‌شده
    base_metrics  = eval_with_threshold(y_test, prob_test, 0.5)
    tuned_metrics = eval_with_threshold(y_test, prob_test, tuned_t)

    auc = roc_auc_score(y_test, prob_test)

    results.append({
        "case": case_id,
        "file": path.split("\\")[-1],
        "rows_model": len(model_df),
        "class1_ratio": class_ratio_1,
        "missing_HR": miss_hr,
        "missing_SpO2": miss_spo2,
        "roc_auc": auc,
        **{f"base_{k}": v for k, v in base_metrics.items()},
        **{f"tuned_{k}": v for k, v in tuned_metrics.items()},
        "status": "ok"
    })

res_df = pd.DataFrame(results)
res_df
```

```python
import numpy as np
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    confusion_matrix, roc_auc_score,
    precision_score, recall_score, f1_score,
    precision_recall_curve
)

BASE_PATH = r"E:\projects\Telemonitoring-Ai-Pipeline\data\uq-vitalsigns\uqvitalsignsdata"

THRESH_SPO2 = 98
HORIZON_SEC = 5 * 60
WIN_SEC = 60

# فقط ستون‌های ضروری را می‌خوانیم تا مشکل ناهماهنگی ستون‌ها کم شود
USECOLS = ["RelativeTimeMilliseconds", "HR", "SpO2"]

def load_case(case_id: int):
    cid = f"case{case_id:02d}"
    path = fr"{BASE_PATH}\{cid}\uq_vsd_{cid}_trenddata.csv"
    df = pd.read_csv(path, engine="python", on_bad_lines="skip", usecols=USECOLS)
    return df, path

def build_dataset(df):
    d = df.copy()

    d["RelativeTimeMilliseconds"] = pd.to_numeric(d["RelativeTimeMilliseconds"], errors="coerce")
    d["HR"] = pd.to_numeric(d["HR"], errors="coerce")
    d["SpO2"] = pd.to_numeric(d["SpO2"], errors="coerce")

    d = d.dropna(subset=["RelativeTimeMilliseconds"]).sort_values("RelativeTimeMilliseconds")

    dt_ms = d["RelativeTimeMilliseconds"].diff().median()
    if pd.isna(dt_ms) or dt_ms <= 0:
        return None

    h = int((HORIZON_SEC * 1000) / dt_ms)
    w = int((WIN_SEC * 1000) / dt_ms)

    # برچسب: آیا در ۵ دقیقه آینده SpO2 < 98 می‌شود؟
    future_min_spo2 = d["SpO2"][::-1].rolling(window=h, min_periods=1).min()[::-1]
    d["y"] = (future_min_spo2 < THRESH_SPO2).astype(int)

    # ویژگی‌ها از پنجره ۶۰ ثانیه گذشته
    d["HR_mean"] = d["HR"].rolling(w, min_periods=w).mean()
    d["HR_std"]  = d["HR"].rolling(w, min_periods=w).std()
    d["HR_min"]  = d["HR"].rolling(w, min_periods=w).min()
    d["HR_max"]  = d["HR"].rolling(w, min_periods=w).max()

    d["SpO2_mean"] = d["SpO2"].rolling(w, min_periods=w).mean()
    d["SpO2_std"]  = d["SpO2"].rolling(w, min_periods=w).std()
    d["SpO2_min"]  = d["SpO2"].rolling(w, min_periods=w).min()
    d["SpO2_max"]  = d["SpO2"].rolling(w, min_periods=w).max()

    d["HR_slope"] = (d["HR"] - d["HR"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))
    d["SpO2_slope"] = (d["SpO2"] - d["SpO2"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))

    feature_cols = ["HR_mean","HR_std","HR_min","HR_max","HR_slope",
                    "SpO2_mean","SpO2_std","SpO2_min","SpO2_max","SpO2_slope"]

    model_df = d.dropna(subset=feature_cols + ["y"]).copy()
    return model_df, feature_cols

def tune_threshold_by_recall(y_true, y_prob, target_recall=0.90):
    prec, rec, thr = precision_recall_curve(y_true, y_prob)
    pr_df = pd.DataFrame({"threshold": np.r_[thr, np.nan], "precision": prec, "recall": rec})
    c = pr_df.dropna(subset=["threshold"])
    c = c[c["recall"] >= target_recall]
    if len(c) == 0:
        return 0.5
    best = c.sort_values(["precision", "recall"], ascending=False).iloc[0]
    return float(best["threshold"])

def eval_at_threshold(y_true, y_prob, t):
    y_pred = (y_prob >= t).astype(int)
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    return {
        "threshold": t,
        "precision_1": precision_score(y_true, y_pred, zero_division=0),
        "recall_1": recall_score(y_true, y_pred, zero_division=0),
        "f1_1": f1_score(y_true, y_pred, zero_division=0),
        "TN": tn, "FP": fp, "FN": fn, "TP": tp
    }

CASE_LIST = [1, 2, 3, 4, 5]

# آماده‌سازی همه کیس‌ها
case_data = {}
feature_cols_global = None

for cid in CASE_LIST:
    df, path = load_case(cid)
    built = build_dataset(df)
    if built is None:
        case_data[cid] = {"status": "build_failed", "path": path}
        continue
    model_df, feature_cols = built
    feature_cols_global = feature_cols
    case_data[cid] = {"status": "ok", "path": path, "model_df": model_df}

results = []

for test_case in CASE_LIST:
    if case_data.get(test_case, {}).get("status") != "ok":
        results.append({"test_case": test_case, "status": "skip_build_failed"})
        continue

    # داده آزمون = کل کیس آزمون
    test_df = case_data[test_case]["model_df"]
    X_test = test_df[feature_cols_global]
    y_test = test_df["y"]

    # داده آموزش = ترکیب بقیه کیس‌ها
    train_parts = []
    for train_case in CASE_LIST:
        if train_case == test_case:
            continue
        if case_data.get(train_case, {}).get("status") == "ok":
            train_parts.append(case_data[train_case]["model_df"])

    if len(train_parts) == 0:
        results.append({"test_case": test_case, "status": "no_train_data"})
        continue

    train_df = pd.concat(train_parts, ignore_index=True)

    X_train = train_df[feature_cols_global]
    y_train = train_df["y"]

    # اگر آموزش تک‌کلاسه شد، این تکرار را رد می‌کنیم
    if y_train.nunique() < 2:
        results.append({"test_case": test_case, "status": "one_class_train"})
        continue

    clf = Pipeline(steps=[
        ("scaler", StandardScaler()),
        ("lr", LogisticRegression(max_iter=2000, class_weight="balanced", random_state=42))
    ])
    clf.fit(X_train, y_train)

    prob_train = clf.predict_proba(X_train)[:, 1]
    prob_test = clf.predict_proba(X_test)[:, 1]

    auc = roc_auc_score(y_test, prob_test) if y_test.nunique() == 2 else np.nan

    tuned_t = tune_threshold_by_recall(y_train, prob_train, target_recall=0.90)

    base_metrics = eval_at_threshold(y_test, prob_test, 0.5) if y_test.nunique() == 2 else None
    tuned_metrics = eval_at_threshold(y_test, prob_test, tuned_t) if y_test.nunique() == 2 else None

    row = {
        "test_case": test_case,
        "status": "ok",
        "rows_test": len(test_df),
        "ratio_event_test": float(y_test.mean()),
        "roc_auc": float(auc) if not np.isnan(auc) else np.nan,
        "tuned_threshold": tuned_t,
    }

    if base_metrics:
        row.update({f"base_{k}": v for k, v in base_metrics.items()})
    if tuned_metrics:
        row.update({f"tuned_{k}": v for k, v in tuned_metrics.items()})

    results.append(row)

res_df = pd.DataFrame(results)
res_df
```

```python
import numpy as np
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    confusion_matrix, roc_auc_score,
    precision_score, recall_score, f1_score,
    precision_recall_curve
)

BASE_PATH = r"E:\projects\Telemonitoring-Ai-Pipeline\data\uq-vitalsigns\uqvitalsignsdata"

THRESH_SPO2 = 98
HORIZON_SEC = 5 * 60
WIN_SEC = 60

def load_case_full(case_id: int):
    cid = f"case{case_id:02d}"
    path = fr"{BASE_PATH}\{cid}\uq_vsd_{cid}_trenddata.csv"
    df = pd.read_csv(path, sep=",", header=0, engine="python", on_bad_lines="skip", index_col=False)
    return df, path

def extract_and_resample(df):
    # فقط ستون‌های مورد نیاز را جدا می‌کنیم
    needed = ["RelativeTimeMilliseconds", "HR", "SpO2"]
    for c in needed:
        if c not in df.columns:
            return None

    d = df[needed].copy()
    d["RelativeTimeMilliseconds"] = pd.to_numeric(d["RelativeTimeMilliseconds"], errors="coerce")
    d["HR"] = pd.to_numeric(d["HR"], errors="coerce")
    d["SpO2"] = pd.to_numeric(d["SpO2"], errors="coerce")
    d = d.dropna(subset=["RelativeTimeMilliseconds"]).sort_values("RelativeTimeMilliseconds")

    if len(d) < 200:
        return None

    # تخمین گام زمانی (فقط فاصله‌های مثبت)
    diffs = d["RelativeTimeMilliseconds"].diff()
    dt_ms = diffs[diffs > 0].median()
    if pd.isna(dt_ms) or dt_ms <= 0:
        return None

    # اگر نرخ نمونه‌برداری خیلی ریز بود، به ۱ ثانیه‌ای تبدیل می‌کنیم
    # معیار: اگر dt_ms خیلی کوچک‌تر از 500 بود، یعنی داده بسیار پرتراکم است
    if dt_ms < 500:
        d["t_bin"] = (d["RelativeTimeMilliseconds"] // 1000) * 1000
        d = d.groupby("t_bin", as_index=False).agg({
            "RelativeTimeMilliseconds": "first",
            "HR": "mean",
            "SpO2": "mean"
        })
        d = d.sort_values("RelativeTimeMilliseconds")
        # دوباره dt را آپدیت می‌کنیم
        diffs = d["RelativeTimeMilliseconds"].diff()
        dt_ms = diffs[diffs > 0].median()

    return d, float(dt_ms)

def build_dataset(d, dt_ms):
    h = int((HORIZON_SEC * 1000) / dt_ms)
    w = int((WIN_SEC * 1000) / dt_ms)

    # برچسب: آیا در ۵ دقیقه آینده SpO2 < 98 می‌شود؟
    future_min_spo2 = d["SpO2"][::-1].rolling(window=h, min_periods=1).min()[::-1]
    d = d.copy()
    d["y"] = (future_min_spo2 < THRESH_SPO2).astype(int)

    # ویژگی‌ها (۶۰ ثانیه گذشته)
    d["HR_mean"] = d["HR"].rolling(w, min_periods=w).mean()
    d["HR_std"]  = d["HR"].rolling(w, min_periods=w).std()
    d["HR_min"]  = d["HR"].rolling(w, min_periods=w).min()
    d["HR_max"]  = d["HR"].rolling(w, min_periods=w).max()

    d["SpO2_mean"] = d["SpO2"].rolling(w, min_periods=w).mean()
    d["SpO2_std"]  = d["SpO2"].rolling(w, min_periods=w).std()
    d["SpO2_min"]  = d["SpO2"].rolling(w, min_periods=w).min()
    d["SpO2_max"]  = d["SpO2"].rolling(w, min_periods=w).max()

    d["HR_slope"] = (d["HR"] - d["HR"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))
    d["SpO2_slope"] = (d["SpO2"] - d["SpO2"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))

    feature_cols = ["HR_mean","HR_std","HR_min","HR_max","HR_slope",
                    "SpO2_mean","SpO2_std","SpO2_min","SpO2_max","SpO2_slope"]

    model_df = d.dropna(subset=feature_cols + ["y"]).copy()
    return model_df, feature_cols

def tune_threshold_by_recall(y_true, y_prob, target_recall=0.90):
    prec, rec, thr = precision_recall_curve(y_true, y_prob)
    pr_df = pd.DataFrame({"threshold": np.r_[thr, np.nan], "precision": prec, "recall": rec})
    c = pr_df.dropna(subset=["threshold"])
    c = c[c["recall"] >= target_recall]
    if len(c) == 0:
        return 0.5
    best = c.sort_values(["precision", "recall"], ascending=False).iloc[0]
    return float(best["threshold"])

def eval_at_threshold(y_true, y_prob, t):
    y_pred = (y_prob >= t).astype(int)
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    return {
        "threshold": t,
        "precision_1": precision_score(y_true, y_pred, zero_division=0),
        "recall_1": recall_score(y_true, y_pred, zero_division=0),
        "f1_1": f1_score(y_true, y_pred, zero_division=0),
        "TN": tn, "FP": fp, "FN": fn, "TP": tp
    }

CASE_LIST = [1,2,3,4,5]

# آماده‌سازی کیس‌ها
case_data = {}
feature_cols_global = None

for cid in CASE_LIST:
    df, path = load_case_full(cid)
    ext = extract_and_resample(df)
    if ext is None:
        case_data[cid] = {"status": "build_failed", "path": path}
        continue
    d, dt_ms = ext
    built = build_dataset(d, dt_ms)
    if built is None:
        case_data[cid] = {"status": "build_failed", "path": path}
        continue
    model_df, feature_cols = built
    feature_cols_global = feature_cols
    case_data[cid] = {"status": "ok", "path": path, "model_df": model_df}

results = []

for test_case in CASE_LIST:
    if case_data.get(test_case, {}).get("status") != "ok":
        results.append({"test_case": test_case, "status": "skip_build_failed"})
        continue

    test_df = case_data[test_case]["model_df"]
    X_test = test_df[feature_cols_global]
    y_test = test_df["y"]

    train_parts = []
    for train_case in CASE_LIST:
        if train_case == test_case:
            continue
        if case_data.get(train_case, {}).get("status") == "ok":
            train_parts.append(case_data[train_case]["model_df"])

    if len(train_parts) == 0:
        results.append({"test_case": test_case, "status": "no_train_data"})
        continue

    train_df = pd.concat(train_parts, ignore_index=True)
    X_train = train_df[feature_cols_global]
    y_train = train_df["y"]

    if y_train.nunique() < 2:
        results.append({"test_case": test_case, "status": "one_class_train"})
        continue

    clf = Pipeline(steps=[
        ("scaler", StandardScaler()),
        ("lr", LogisticRegression(max_iter=2000, class_weight="balanced", random_state=42))
    ])
    clf.fit(X_train, y_train)

    prob_train = clf.predict_proba(X_train)[:, 1]
    prob_test = clf.predict_proba(X_test)[:, 1]

    auc = roc_auc_score(y_test, prob_test) if y_test.nunique() == 2 else np.nan

    tuned_t = tune_threshold_by_recall(y_train, prob_train, target_recall=0.90)
    base_metrics = eval_at_threshold(y_test, prob_test, 0.5) if y_test.nunique() == 2 else None
    tuned_metrics = eval_at_threshold(y_test, prob_test, tuned_t) if y_test.nunique() == 2 else None

    row = {
        "test_case": test_case,
        "status": "ok",
        "rows_test": len(test_df),
        "ratio_event_test": float(y_test.mean()),
        "roc_auc": float(auc) if not np.isnan(auc) else np.nan,
        "tuned_threshold": tuned_t,
    }
    if base_metrics:
        row.update({f"base_{k}": v for k, v in base_metrics.items()})
    if tuned_metrics:
        row.update({f"tuned_{k}": v for k, v in tuned_metrics.items()})

    results.append(row)

res_df = pd.DataFrame(results)
res_df
```

```python
import numpy as np
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    confusion_matrix, roc_auc_score,
    precision_score, recall_score, f1_score,
    precision_recall_curve
)

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression


# -----------------------------
# تنظیمات کلی
# -----------------------------
BASE_PATH = r"E:\projects\Telemonitoring-Ai-Pipeline\data\uq-vitalsigns\uqvitalsignsdata"
THRESH_SPO2 = 98
HORIZON_SEC = 5 * 60
WIN_SEC = 60

CASE_LIST = [1, 2, 3, 4, 5]   # فعلاً ۵ کیس اول


# -----------------------------
# توابع کمکی
# -----------------------------
def load_case_full(case_id: int):
    cid = f"case{case_id:02d}"
    path = fr"{BASE_PATH}\{cid}\uq_vsd_{cid}_trenddata.csv"
    # خواندن کامل (به‌خاطر بعضی ناسازگاری‌های سطری)
    df = pd.read_csv(path, sep=",", header=0, engine="python", on_bad_lines="skip", index_col=False)
    return df, path


def extract_and_resample(df):
    # ستون‌های لازم
    needed = ["RelativeTimeMilliseconds", "HR", "SpO2"]
    if not all(c in df.columns for c in needed):
        return None

    d = df[needed].copy()
    d["RelativeTimeMilliseconds"] = pd.to_numeric(d["RelativeTimeMilliseconds"], errors="coerce")
    d["HR"] = pd.to_numeric(d["HR"], errors="coerce")
    d["SpO2"] = pd.to_numeric(d["SpO2"], errors="coerce")
    d = d.dropna(subset=["RelativeTimeMilliseconds"]).sort_values("RelativeTimeMilliseconds")

    if len(d) < 200:
        return None

    # گام زمانی از فاصله‌های مثبت
    diffs = d["RelativeTimeMilliseconds"].diff()
    dt_ms = diffs[diffs > 0].median()
    if pd.isna(dt_ms) or dt_ms <= 0:
        return None

    # اگر داده خیلی پرتراکم باشد، به ۱ ثانیه‌ای تبدیل می‌کنیم
    if dt_ms < 500:
        d["t_bin"] = (d["RelativeTimeMilliseconds"] // 1000) * 1000
        d = d.groupby("t_bin", as_index=False).agg({
            "RelativeTimeMilliseconds": "first",
            "HR": "mean",
            "SpO2": "mean"
        }).sort_values("RelativeTimeMilliseconds")

        diffs = d["RelativeTimeMilliseconds"].diff()
        dt_ms = diffs[diffs > 0].median()

    return d, float(dt_ms)


def build_dataset(d, dt_ms):
    h = int((HORIZON_SEC * 1000) / dt_ms)
    w = int((WIN_SEC * 1000) / dt_ms)

    # برچسب: آیا در ۵ دقیقه آینده اکسیژن < ۹۸ می‌شود؟
    future_min_spo2 = d["SpO2"][::-1].rolling(window=h, min_periods=1).min()[::-1]
    d = d.copy()
    d["y"] = (future_min_spo2 < THRESH_SPO2).astype(int)

    # ویژگی‌ها از پنجره ۶۰ ثانیه گذشته
    d["HR_mean"] = d["HR"].rolling(w, min_periods=w).mean()
    d["HR_std"]  = d["HR"].rolling(w, min_periods=w).std()
    d["HR_min"]  = d["HR"].rolling(w, min_periods=w).min()
    d["HR_max"]  = d["HR"].rolling(w, min_periods=w).max()

    d["SpO2_mean"] = d["SpO2"].rolling(w, min_periods=w).mean()
    d["SpO2_std"]  = d["SpO2"].rolling(w, min_periods=w).std()
    d["SpO2_min"]  = d["SpO2"].rolling(w, min_periods=w).min()
    d["SpO2_max"]  = d["SpO2"].rolling(w, min_periods=w).max()

    d["HR_slope"] = (d["HR"] - d["HR"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))
    d["SpO2_slope"] = (d["SpO2"] - d["SpO2"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))

    feature_cols = ["HR_mean","HR_std","HR_min","HR_max","HR_slope",
                    "SpO2_mean","SpO2_std","SpO2_min","SpO2_max","SpO2_slope"]

    model_df = d.dropna(subset=feature_cols + ["y"]).copy()
    return model_df, feature_cols


def tune_threshold_by_recall(y_true, y_prob, target_recall=0.90):
    # تنظیم آستانه با شرط فراخوانی (بر اساس منحنی دقت–فراخوانی) 【1-c51a64】
    prec, rec, thr = precision_recall_curve(y_true, y_prob)
    pr_df = pd.DataFrame({"threshold": np.r_[thr, np.nan], "precision": prec, "recall": rec})
    c = pr_df.dropna(subset=["threshold"])
    c = c[c["recall"] >= target_recall]
    if len(c) == 0:
        return 0.5
    best = c.sort_values(["precision", "recall"], ascending=False).iloc[0]
    return float(best["threshold"])


def eval_at_threshold(y_true, y_prob, t):
    y_pred = (y_prob >= t).astype(int)
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    return {
        "threshold": t,
        "precision_1": precision_score(y_true, y_pred, zero_division=0),
        "recall_1": recall_score(y_true, y_pred, zero_division=0),
        "f1_1": f1_score(y_true, y_pred, zero_division=0),
        "TN": tn, "FP": fp, "FN": fn, "TP": tp
    }


# -----------------------------
# تعریف دو مدل (به همراه مدل پایه قبلی برای مقایسه)
# -----------------------------
def make_models():
    models = {}

    # مدل پایه قبلی (برای اینکه ببینیم بهتر شدیم یا نه)
    models["پایه_رگرسیون"] = Pipeline(steps=[
        ("scaler", StandardScaler()),
        ("clf", LogisticRegression(max_iter=2000, class_weight="balanced", random_state=42))
    ])

    # مدل ۱: جنگل تصادفی
    models["جنگل_تصادفی"] = RandomForestClassifier(
        n_estimators=300,
        max_depth=None,
        min_samples_leaf=2,
        class_weight="balanced",
        random_state=42,
        n_jobs=-1
    )

    # مدل ۲: تقویتی گرادیانی
    models["تقویتی_گرادیانی"] = GradientBoostingClassifier(
        random_state=42
    )

    return models


# -----------------------------
# آماده‌سازی کیس‌ها (یک‌بار)
# -----------------------------
case_data = {}
feature_cols_global = None

for cid in CASE_LIST:
    df, path = load_case_full(cid)
    ext = extract_and_resample(df)
    if ext is None:
        case_data[cid] = {"status": "build_failed", "path": path}
        continue
    d, dt_ms = ext
    model_df, feature_cols = build_dataset(d, dt_ms)
    feature_cols_global = feature_cols
    case_data[cid] = {"status": "ok", "path": path, "model_df": model_df}

models = make_models()

# -----------------------------
# ارزیابی: هر بار یک کیس آزمون، بقیه آموزش
# -----------------------------
rows = []

for test_case in CASE_LIST:
    if case_data.get(test_case, {}).get("status") != "ok":
        rows.append({"test_case": test_case, "status": "skip_build_failed"})
        continue

    test_df = case_data[test_case]["model_df"]
    X_test = test_df[feature_cols_global]
    y_test = test_df["y"].astype(int)

    train_parts = []
    for train_case in CASE_LIST:
        if train_case == test_case:
            continue
        if case_data.get(train_case, {}).get("status") == "ok":
            train_parts.append(case_data[train_case]["model_df"])

    if len(train_parts) == 0:
        rows.append({"test_case": test_case, "status": "no_train_data"})
        continue

    train_df = pd.concat(train_parts, ignore_index=True)
    X_train = train_df[feature_cols_global]
    y_train = train_df["y"].astype(int)

    # اگر آزمون یا آموزش تک‌کلاسه باشد، ارزیابی معنی‌دار نیست
    if y_train.nunique() < 2 or y_test.nunique() < 2:
        rows.append({"test_case": test_case, "status": "one_class_issue"})
        continue

    for name, model in models.items():
        # آموزش
        model_fit = model
        model_fit.fit(X_train, y_train)

        # احتمال‌ها
        if hasattr(model_fit, "predict_proba"):
            prob_train = model_fit.predict_proba(X_train)[:, 1]
            prob_test  = model_fit.predict_proba(X_test)[:, 1]
        else:
            # اگر مدل احتمال نداد، این مورد را رد می‌کنیم
            rows.append({"test_case": test_case, "model": name, "status": "no_proba"})
            continue

        auc = roc_auc_score(y_test, prob_test)

        # آستانه را فقط با آموزش تنظیم می‌کنیم (هدف فراخوانی ۰.۹۰)
        tuned_t = tune_threshold_by_recall(y_train, prob_train, target_recall=0.90)

        # ارزیابی با آستانه پیش‌فرض و آستانه تنظیم‌شده
        base_m  = eval_at_threshold(y_test, prob_test, 0.5)
        tuned_m = eval_at_threshold(y_test, prob_test, tuned_t)

        row = {
            "test_case": test_case,
            "model": name,
            "status": "ok",
            "roc_auc": auc,
            "tuned_threshold": tuned_t,
            **{f"base_{k}": v for k, v in base_m.items()},
            **{f"tuned_{k}": v for k, v in tuned_m.items()},
        }
        rows.append(row)

compare_df = pd.DataFrame(rows)
compare_df
```

```python
import numpy as np
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    confusion_matrix, roc_auc_score,
    precision_score, recall_score, f1_score,
    precision_recall_curve
)

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression


# -----------------------------
# تنظیمات کلی
# -----------------------------
BASE_PATH = r"E:\projects\Telemonitoring-Ai-Pipeline\data\uq-vitalsigns\uqvitalsignsdata"
THRESH_SPO2 = 98
HORIZON_SEC = 5 * 60
WIN_SEC = 60

CASE_LIST = [1, 2, 3, 4, 5]   # فعلاً ۵ کیس اول


# -----------------------------
# توابع کمکی
# -----------------------------
def load_case_full(case_id: int):
    cid = f"case{case_id:02d}"
    path = fr"{BASE_PATH}\{cid}\uq_vsd_{cid}_trenddata.csv"
    # خواندن کامل (به‌خاطر بعضی ناسازگاری‌های سطری)
    df = pd.read_csv(path, sep=",", header=0, engine="python", on_bad_lines="skip", index_col=False)
    return df, path


def extract_and_resample(df):
    # ستون‌های لازم
    needed = ["RelativeTimeMilliseconds", "HR", "SpO2"]
    if not all(c in df.columns for c in needed):
        return None

    d = df[needed].copy()
    d["RelativeTimeMilliseconds"] = pd.to_numeric(d["RelativeTimeMilliseconds"], errors="coerce")
    d["HR"] = pd.to_numeric(d["HR"], errors="coerce")
    d["SpO2"] = pd.to_numeric(d["SpO2"], errors="coerce")
    d = d.dropna(subset=["RelativeTimeMilliseconds"]).sort_values("RelativeTimeMilliseconds")

    if len(d) < 200:
        return None

    # گام زمانی از فاصله‌های مثبت
    diffs = d["RelativeTimeMilliseconds"].diff()
    dt_ms = diffs[diffs > 0].median()
    if pd.isna(dt_ms) or dt_ms <= 0:
        return None

    # اگر داده خیلی پرتراکم باشد، به ۱ ثانیه‌ای تبدیل می‌کنیم
    if dt_ms < 500:
        d["t_bin"] = (d["RelativeTimeMilliseconds"] // 1000) * 1000
        d = d.groupby("t_bin", as_index=False).agg({
            "RelativeTimeMilliseconds": "first",
            "HR": "mean",
            "SpO2": "mean"
        }).sort_values("RelativeTimeMilliseconds")

        diffs = d["RelativeTimeMilliseconds"].diff()
        dt_ms = diffs[diffs > 0].median()

    return d, float(dt_ms)


def build_dataset(d, dt_ms):
    h = int((HORIZON_SEC * 1000) / dt_ms)
    w = int((WIN_SEC * 1000) / dt_ms)

    # برچسب: آیا در ۵ دقیقه آینده اکسیژن < ۹۸ می‌شود؟
    future_min_spo2 = d["SpO2"][::-1].rolling(window=h, min_periods=1).min()[::-1]
    d = d.copy()
    d["y"] = (future_min_spo2 < THRESH_SPO2).astype(int)

    # ویژگی‌ها از پنجره ۶۰ ثانیه گذشته
    d["HR_mean"] = d["HR"].rolling(w, min_periods=w).mean()
    d["HR_std"]  = d["HR"].rolling(w, min_periods=w).std()
    d["HR_min"]  = d["HR"].rolling(w, min_periods=w).min()
    d["HR_max"]  = d["HR"].rolling(w, min_periods=w).max()

    d["SpO2_mean"] = d["SpO2"].rolling(w, min_periods=w).mean()
    d["SpO2_std"]  = d["SpO2"].rolling(w, min_periods=w).std()
    d["SpO2_min"]  = d["SpO2"].rolling(w, min_periods=w).min()
    d["SpO2_max"]  = d["SpO2"].rolling(w, min_periods=w).max()

    d["HR_slope"] = (d["HR"] - d["HR"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))
    d["SpO2_slope"] = (d["SpO2"] - d["SpO2"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))

    feature_cols = ["HR_mean","HR_std","HR_min","HR_max","HR_slope",
                    "SpO2_mean","SpO2_std","SpO2_min","SpO2_max","SpO2_slope"]

    model_df = d.dropna(subset=feature_cols + ["y"]).copy()
    return model_df, feature_cols


def tune_threshold_by_recall(y_true, y_prob, target_recall=0.80):
    # تنظیم آستانه با شرط فراخوانی (بر اساس منحنی دقت–فراخوانی) 【1-c51a64】
    prec, rec, thr = precision_recall_curve(y_true, y_prob)
    pr_df = pd.DataFrame({"threshold": np.r_[thr, np.nan], "precision": prec, "recall": rec})
    c = pr_df.dropna(subset=["threshold"])
    c = c[c["recall"] >= target_recall]
    if len(c) == 0:
        return 0.5
    best = c.sort_values(["precision", "recall"], ascending=False).iloc[0]
    return float(best["threshold"])


def eval_at_threshold(y_true, y_prob, t):
    y_pred = (y_prob >= t).astype(int)
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    return {
        "threshold": t,
        "precision_1": precision_score(y_true, y_pred, zero_division=0),
        "recall_1": recall_score(y_true, y_pred, zero_division=0),
        "f1_1": f1_score(y_true, y_pred, zero_division=0),
        "TN": tn, "FP": fp, "FN": fn, "TP": tp
    }


# -----------------------------
# تعریف دو مدل (به همراه مدل پایه قبلی برای مقایسه)
# -----------------------------
def make_models():
    models = {}

    # مدل پایه قبلی (برای اینکه ببینیم بهتر شدیم یا نه)
    models["پایه_رگرسیون"] = Pipeline(steps=[
        ("scaler", StandardScaler()),
        ("clf", LogisticRegression(max_iter=2000, class_weight="balanced", random_state=42))
    ])

    # مدل ۱: جنگل تصادفی
    models["جنگل_تصادفی"] = RandomForestClassifier(
        n_estimators=300,
        max_depth=None,
        min_samples_leaf=2,
        class_weight="balanced",
        random_state=42,
        n_jobs=-1
    )

    # مدل ۲: تقویتی گرادیانی
    models["تقویتی_گرادیانی"] = GradientBoostingClassifier(
        random_state=42
    )

    return models


# -----------------------------
# آماده‌سازی کیس‌ها (یک‌بار)
# -----------------------------
case_data = {}
feature_cols_global = None

for cid in CASE_LIST:
    df, path = load_case_full(cid)
    ext = extract_and_resample(df)
    if ext is None:
        case_data[cid] = {"status": "build_failed", "path": path}
        continue
    d, dt_ms = ext
    model_df, feature_cols = build_dataset(d, dt_ms)
    feature_cols_global = feature_cols
    case_data[cid] = {"status": "ok", "path": path, "model_df": model_df}

models = make_models()

# -----------------------------
# ارزیابی: هر بار یک کیس آزمون، بقیه آموزش
# -----------------------------
rows = []

for test_case in CASE_LIST:
    if case_data.get(test_case, {}).get("status") != "ok":
        rows.append({"test_case": test_case, "status": "skip_build_failed"})
        continue

    test_df = case_data[test_case]["model_df"]
    X_test = test_df[feature_cols_global]
    y_test = test_df["y"].astype(int)

    train_parts = []
    for train_case in CASE_LIST:
        if train_case == test_case:
            continue
        if case_data.get(train_case, {}).get("status") == "ok":
            train_parts.append(case_data[train_case]["model_df"])

    if len(train_parts) == 0:
        rows.append({"test_case": test_case, "status": "no_train_data"})
        continue

    train_df = pd.concat(train_parts, ignore_index=True)
    X_train = train_df[feature_cols_global]
    y_train = train_df["y"].astype(int)

    # اگر آزمون یا آموزش تک‌کلاسه باشد، ارزیابی معنی‌دار نیست
    if y_train.nunique() < 2 or y_test.nunique() < 2:
        rows.append({"test_case": test_case, "status": "one_class_issue"})
        continue

    for name, model in models.items():
        # آموزش
        model_fit = model
        model_fit.fit(X_train, y_train)

        # احتمال‌ها
        if hasattr(model_fit, "predict_proba"):
            prob_train = model_fit.predict_proba(X_train)[:, 1]
            prob_test  = model_fit.predict_proba(X_test)[:, 1]
        else:
            # اگر مدل احتمال نداد، این مورد را رد می‌کنیم
            rows.append({"test_case": test_case, "model": name, "status": "no_proba"})
            continue

        auc = roc_auc_score(y_test, prob_test)

        # آستانه را فقط با آموزش تنظیم می‌کنیم (هدف فراخوانی ۰.۹۰)
        tuned_t = tune_threshold_by_recall(y_train, prob_train, target_recall=0.80)

        # ارزیابی با آستانه پیش‌فرض و آستانه تنظیم‌شده
        base_m  = eval_at_threshold(y_test, prob_test, 0.5)
        tuned_m = eval_at_threshold(y_test, prob_test, tuned_t)

        row = {
            "test_case": test_case,
            "model": name,
            "status": "ok",
            "roc_auc": auc,
            "tuned_threshold": tuned_t,
            **{f"base_{k}": v for k, v in base_m.items()},
            **{f"tuned_{k}": v for k, v in tuned_m.items()},
        }
        rows.append(row)

compare_df = pd.DataFrame(rows)
compare_df
```

```python
import numpy as np
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import (
    confusion_matrix, roc_auc_score,
    precision_score, recall_score, f1_score,
    precision_recall_curve
)

# -----------------------------
# تنظیمات
# -----------------------------
BASE_PATH = r"E:\projects\Telemonitoring-Ai-Pipeline\data\uq-vitalsigns\uqvitalsignsdata"
THRESH_SPO2 = 98
HORIZON_SEC = 5 * 60
WIN_SEC = 60

CASE_LIST = [1,2,3,4,5]

TARGET_RECALL_STAGE1 = 0.80   # هدف فراخوانی مرحله اول
TARGET_RECALL_FINAL  = 0.75   # هدف فراخوانی نهایی (کمی ملایم‌تر تا خیلی سخت‌گیر نشود)


# -----------------------------
# توابع کمکی داده
# -----------------------------
def load_case_full(case_id: int):
    cid = f"case{case_id:02d}"
    path = fr"{BASE_PATH}\{cid}\uq_vsd_{cid}_trenddata.csv"
    df = pd.read_csv(path, sep=",", header=0, engine="python", on_bad_lines="skip", index_col=False)
    return df, path

def extract_and_resample(df):
    needed = ["RelativeTimeMilliseconds", "HR", "SpO2"]
    if not all(c in df.columns for c in needed):
        return None

    d = df[needed].copy()
    d["RelativeTimeMilliseconds"] = pd.to_numeric(d["RelativeTimeMilliseconds"], errors="coerce")
    d["HR"] = pd.to_numeric(d["HR"], errors="coerce")
    d["SpO2"] = pd.to_numeric(d["SpO2"], errors="coerce")
    d = d.dropna(subset=["RelativeTimeMilliseconds"]).sort_values("RelativeTimeMilliseconds")

    if len(d) < 200:
        return None

    diffs = d["RelativeTimeMilliseconds"].diff()
    dt_ms = diffs[diffs > 0].median()
    if pd.isna(dt_ms) or dt_ms <= 0:
        return None

    # اگر داده خیلی پرتراکم باشد -> به ۱ ثانیه‌ای تبدیل
    if dt_ms < 500:
        d["t_bin"] = (d["RelativeTimeMilliseconds"] // 1000) * 1000
        d = d.groupby("t_bin", as_index=False).agg({
            "RelativeTimeMilliseconds": "first",
            "HR": "mean",
            "SpO2": "mean"
        }).sort_values("RelativeTimeMilliseconds")

        diffs = d["RelativeTimeMilliseconds"].diff()
        dt_ms = diffs[diffs > 0].median()

    return d, float(dt_ms)

def build_dataset(d, dt_ms):
    h = int((HORIZON_SEC * 1000) / dt_ms)
    w = int((WIN_SEC * 1000) / dt_ms)

    future_min_spo2 = d["SpO2"][::-1].rolling(window=h, min_periods=1).min()[::-1]
    d = d.copy()
    d["y"] = (future_min_spo2 < THRESH_SPO2).astype(int)

    d["HR_mean"] = d["HR"].rolling(w, min_periods=w).mean()
    d["HR_std"]  = d["HR"].rolling(w, min_periods=w).std()
    d["HR_min"]  = d["HR"].rolling(w, min_periods=w).min()
    d["HR_max"]  = d["HR"].rolling(w, min_periods=w).max()

    d["SpO2_mean"] = d["SpO2"].rolling(w, min_periods=w).mean()
    d["SpO2_std"]  = d["SpO2"].rolling(w, min_periods=w).std()
    d["SpO2_min"]  = d["SpO2"].rolling(w, min_periods=w).min()
    d["SpO2_max"]  = d["SpO2"].rolling(w, min_periods=w).max()

    d["HR_slope"] = (d["HR"] - d["HR"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))
    d["SpO2_slope"] = (d["SpO2"] - d["SpO2"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))

    feature_cols = ["HR_mean","HR_std","HR_min","HR_max","HR_slope",
                    "SpO2_mean","SpO2_std","SpO2_min","SpO2_max","SpO2_slope"]

    model_df = d.dropna(subset=feature_cols + ["y"]).copy()
    return model_df, feature_cols


# -----------------------------
# تنظیم آستانه بر اساس هدف فراخوانی
# -----------------------------
def tune_threshold_by_recall(y_true, y_prob, target_recall):
    prec, rec, thr = precision_recall_curve(y_true, y_prob)
    pr_df = pd.DataFrame({"threshold": np.r_[thr, np.nan], "precision": prec, "recall": rec})
    c = pr_df.dropna(subset=["threshold"])
    c = c[c["recall"] >= target_recall]
    if len(c) == 0:
        return 0.5
    # بین مواردی که فراخوانی کافی دارند، بیشترین دقت را انتخاب می‌کنیم
    best = c.sort_values(["precision", "recall"], ascending=False).iloc[0]
    return float(best["threshold"])


# -----------------------------
# ارزیابی یک پیش‌بینی
# -----------------------------
def eval_pred(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    return {
        "precision_1": precision_score(y_true, y_pred, zero_division=0),
        "recall_1": recall_score(y_true, y_pred, zero_division=0),
        "f1_1": f1_score(y_true, y_pred, zero_division=0),
        "TN": tn, "FP": fp, "FN": fn, "TP": tp
    }


# -----------------------------
# انتخاب آستانه مرحله دوم (ملایم) با حفظ فراخوانی نهایی
# -----------------------------
def tune_stage2_threshold_soft(y_train, prob1_train, t1, prob2_train, target_final_recall):
    """
    روی داده آموزش:
    - اول با آستانه مرحله اول t1 نمونه‌های مشکوک را انتخاب می‌کنیم.
    - سپس آستانه مرحله دوم را طوری انتخاب می‌کنیم که فراخوانی نهایی (پس از دو مرحله) حداقل به target_final_recall برسد
      و در بین آن‌ها دقت نهایی بیشینه شود.
    """
    cand_mask = (prob1_train >= t1)
    if cand_mask.sum() == 0:
        return 0.5

    # فقط آستانه‌هایی را بررسی می‌کنیم که از خروجی مرحله دوم می‌آید
    prec2, rec2, thr2 = precision_recall_curve(y_train[cand_mask], prob2_train[cand_mask])

    best_t2 = 0.5
    best_prec = -1

    for t2 in thr2:
        y_pred_final = (prob1_train >= t1) & (prob2_train >= t2)
        y_pred_final = y_pred_final.astype(int)

        r = recall_score(y_train, y_pred_final, zero_division=0)
        p = precision_score(y_train, y_pred_final, zero_division=0)

        if r >= target_final_recall and p > best_prec:
            best_prec = p
            best_t2 = float(t2)

    return best_t2


# -----------------------------
# آماده‌سازی کیس‌ها
# -----------------------------
case_data = {}
feature_cols_global = None

for cid in CASE_LIST:
    df, path = load_case_full(cid)
    ext = extract_and_resample(df)
    if ext is None:
        case_data[cid] = {"status": "build_failed", "path": path}
        continue
    d, dt_ms = ext
    model_df, feature_cols = build_dataset(d, dt_ms)
    feature_cols_global = feature_cols
    case_data[cid] = {"status": "ok", "path": path, "model_df": model_df}


# -----------------------------
# تعریف دو مدل
# -----------------------------
model_stage1 = GradientBoostingClassifier(random_state=42)

model_stage2 = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("clf", LogisticRegression(max_iter=2000, class_weight="balanced", random_state=42))
])


# -----------------------------
# ارزیابی چندکیسی
# -----------------------------
rows = []

for test_case in CASE_LIST:
    if case_data.get(test_case, {}).get("status") != "ok":
        rows.append({"test_case": test_case, "status": "skip_build_failed"})
        continue

    test_df = case_data[test_case]["model_df"]
    X_test = test_df[feature_cols_global]
    y_test = test_df["y"].astype(int)

    train_parts = []
    for train_case in CASE_LIST:
        if train_case == test_case:
            continue
        if case_data.get(train_case, {}).get("status") == "ok":
            train_parts.append(case_data[train_case]["model_df"])

    if len(train_parts) == 0:
        rows.append({"test_case": test_case, "status": "no_train_data"})
        continue

    train_df = pd.concat(train_parts, ignore_index=True)
    X_train = train_df[feature_cols_global]
    y_train = train_df["y"].astype(int)

    if y_train.nunique() < 2 or y_test.nunique() < 2:
        rows.append({"test_case": test_case, "status": "one_class_issue"})
        continue

    # --- آموزش مرحله اول و دوم ---
    model_stage1.fit(X_train, y_train)
    model_stage2.fit(X_train, y_train)

    prob1_train = model_stage1.predict_proba(X_train)[:, 1]
    prob1_test  = model_stage1.predict_proba(X_test)[:, 1]

    prob2_train = model_stage2.predict_proba(X_train)[:, 1]
    prob2_test  = model_stage2.predict_proba(X_test)[:, 1]

    # امتیاز جداسازی کلی (مرحله اول)
    auc1 = roc_auc_score(y_test, prob1_test)

    # --- تنظیم آستانه مرحله اول با هدف ۰٫۸۰ ---
    t1 = tune_threshold_by_recall(y_train, prob1_train, TARGET_RECALL_STAGE1)

    # --- تنظیم آستانه مرحله دوم (ملایم) با حفظ فراخوانی نهایی ---
    t2 = tune_stage2_threshold_soft(y_train, prob1_train, t1, prob2_train, TARGET_RECALL_FINAL)

    # --- سه حالت ارزیابی روی آزمون ---
    # (الف) فقط مرحله اول
    y_pred_s1 = (prob1_test >= t1).astype(int)
    m_s1 = eval_pred(y_test, y_pred_s1)

    # (ب) فقط مرحله دوم
    t2_alone = tune_threshold_by_recall(y_train, prob2_train, TARGET_RECALL_STAGE1)
    y_pred_s2 = (prob2_test >= t2_alone).astype(int)
    m_s2 = eval_pred(y_test, y_pred_s2)

    # (ج) دو مرحله‌ای
    y_pred_2stage = ((prob1_test >= t1) & (prob2_test >= t2)).astype(int)
    m_2s = eval_pred(y_test, y_pred_2stage)

    rows.append({
        "test_case": test_case,
        "status": "ok",
        "roc_auc_stage1": auc1,
        "t1": t1,
        "t2": t2,
        "t2_alone": t2_alone,
        **{f"s1_{k}": v for k, v in m_s1.items()},
        **{f"s2_{k}": v for k, v in m_s2.items()},
        **{f"two_{k}": v for k, v in m_2s.items()},
    })

two_stage_df = pd.DataFrame(rows)
two_stage_df
```

```python
import numpy as np
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import (
    confusion_matrix, roc_auc_score,
    precision_score, recall_score, f1_score,
    precision_recall_curve
)

# -----------------------------
# تنظیمات
# -----------------------------
BASE_PATH = r"E:\projects\Telemonitoring-Ai-Pipeline\data\uq-vitalsigns\uqvitalsignsdata"
THRESH_SPO2 = 98
HORIZON_SEC = 5 * 60
WIN_SEC = 60

CASE_LIST = [1,2,3,4,5]

TARGET_RECALL_STAGE1 = 0.80   # هدف فراخوانی مرحله اول
TARGET_RECALL_FINAL  = 0.75   # هدف فراخوانی نهایی (کمی ملایم‌تر تا خیلی سخت‌گیر نشود)


# -----------------------------
# توابع کمکی داده
# -----------------------------
def load_case_full(case_id: int):
    cid = f"case{case_id:02d}"
    path = fr"{BASE_PATH}\{cid}\uq_vsd_{cid}_trenddata.csv"
    df = pd.read_csv(path, sep=",", header=0, engine="python", on_bad_lines="skip", index_col=False)
    return df, path

def extract_and_resample(df):
    needed = ["RelativeTimeMilliseconds", "HR", "SpO2"]
    if not all(c in df.columns for c in needed):
        return None

    d = df[needed].copy()
    d["RelativeTimeMilliseconds"] = pd.to_numeric(d["RelativeTimeMilliseconds"], errors="coerce")
    d["HR"] = pd.to_numeric(d["HR"], errors="coerce")
    d["SpO2"] = pd.to_numeric(d["SpO2"], errors="coerce")
    d = d.dropna(subset=["RelativeTimeMilliseconds"]).sort_values("RelativeTimeMilliseconds")

    if len(d) < 200:
        return None

    diffs = d["RelativeTimeMilliseconds"].diff()
    dt_ms = diffs[diffs > 0].median()
    if pd.isna(dt_ms) or dt_ms <= 0:
        return None

    # اگر داده خیلی پرتراکم باشد -> به ۱ ثانیه‌ای تبدیل
    if dt_ms < 500:
        d["t_bin"] = (d["RelativeTimeMilliseconds"] // 1000) * 1000
        d = d.groupby("t_bin", as_index=False).agg({
            "RelativeTimeMilliseconds": "first",
            "HR": "mean",
            "SpO2": "mean"
        }).sort_values("RelativeTimeMilliseconds")

        diffs = d["RelativeTimeMilliseconds"].diff()
        dt_ms = diffs[diffs > 0].median()

    return d, float(dt_ms)

def build_dataset(d, dt_ms):
    h = int((HORIZON_SEC * 1000) / dt_ms)
    w = int((WIN_SEC * 1000) / dt_ms)

    future_min_spo2 = d["SpO2"][::-1].rolling(window=h, min_periods=1).min()[::-1]
    d = d.copy()
    d["y"] = (future_min_spo2 < THRESH_SPO2).astype(int)

    d["HR_mean"] = d["HR"].rolling(w, min_periods=w).mean()
    d["HR_std"]  = d["HR"].rolling(w, min_periods=w).std()
    d["HR_min"]  = d["HR"].rolling(w, min_periods=w).min()
    d["HR_max"]  = d["HR"].rolling(w, min_periods=w).max()

    d["SpO2_mean"] = d["SpO2"].rolling(w, min_periods=w).mean()
    d["SpO2_std"]  = d["SpO2"].rolling(w, min_periods=w).std()
    d["SpO2_min"]  = d["SpO2"].rolling(w, min_periods=w).min()
    d["SpO2_max"]  = d["SpO2"].rolling(w, min_periods=w).max()

    d["HR_slope"] = (d["HR"] - d["HR"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))
    d["SpO2_slope"] = (d["SpO2"] - d["SpO2"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))

    feature_cols = ["HR_mean","HR_std","HR_min","HR_max","HR_slope",
                    "SpO2_mean","SpO2_std","SpO2_min","SpO2_max","SpO2_slope"]

    model_df = d.dropna(subset=feature_cols + ["y"]).copy()
    return model_df, feature_cols


# -----------------------------
# تنظیم آستانه بر اساس هدف فراخوانی
# -----------------------------
def tune_threshold_by_recall(y_true, y_prob, target_recall):
    prec, rec, thr = precision_recall_curve(y_true, y_prob)
    pr_df = pd.DataFrame({"threshold": np.r_[thr, np.nan], "precision": prec, "recall": rec})
    c = pr_df.dropna(subset=["threshold"])
    c = c[c["recall"] >= target_recall]
    if len(c) == 0:
        return 0.5
    # بین مواردی که فراخوانی کافی دارند، بیشترین دقت را انتخاب می‌کنیم
    best = c.sort_values(["precision", "recall"], ascending=False).iloc[0]
    return float(best["threshold"])


# -----------------------------
# ارزیابی یک پیش‌بینی
# -----------------------------
def eval_pred(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    return {
        "precision_1": precision_score(y_true, y_pred, zero_division=0),
        "recall_1": recall_score(y_true, y_pred, zero_division=0),
        "f1_1": f1_score(y_true, y_pred, zero_division=0),
        "TN": tn, "FP": fp, "FN": fn, "TP": tp
    }


# -----------------------------
# انتخاب آستانه مرحله دوم (ملایم) با حفظ فراخوانی نهایی
# -----------------------------
def tune_stage2_threshold_soft(y_train, prob1_train, t1, prob2_train, target_final_recall):
    """
    روی داده آموزش:
    - اول با آستانه مرحله اول t1 نمونه‌های مشکوک را انتخاب می‌کنیم.
    - سپس آستانه مرحله دوم را طوری انتخاب می‌کنیم که فراخوانی نهایی (پس از دو مرحله) حداقل به target_final_recall برسد
      و در بین آن‌ها دقت نهایی بیشینه شود.
    """
    cand_mask = (prob1_train >= t1)
    if cand_mask.sum() == 0:
        return 0.5

    # فقط آستانه‌هایی را بررسی می‌کنیم که از خروجی مرحله دوم می‌آید
    prec2, rec2, thr2 = precision_recall_curve(y_train[cand_mask], prob2_train[cand_mask])

    best_t2 = 0.5
    best_prec = -1

    for t2 in thr2:
        y_pred_final = (prob1_train >= t1) & (prob2_train >= t2)
        y_pred_final = y_pred_final.astype(int)

        r = recall_score(y_train, y_pred_final, zero_division=0)
        p = precision_score(y_train, y_pred_final, zero_division=0)

        if r >= target_final_recall and p > best_prec:
            best_prec = p
            best_t2 = float(t2)

    return best_t2


# -----------------------------
# آماده‌سازی کیس‌ها
# -----------------------------
case_data = {}
feature_cols_global = None

for cid in CASE_LIST:
    df, path = load_case_full(cid)
    ext = extract_and_resample(df)
    if ext is None:
        case_data[cid] = {"status": "build_failed", "path": path}
        continue
    d, dt_ms = ext
    model_df, feature_cols = build_dataset(d, dt_ms)
    feature_cols_global = feature_cols
    case_data[cid] = {"status": "ok", "path": path, "model_df": model_df}


# -----------------------------
# تعریف دو مدل
# -----------------------------
model_stage1 = GradientBoostingClassifier(random_state=42)

model_stage2 = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("clf", LogisticRegression(max_iter=2000, class_weight="balanced", random_state=42))
])


# -----------------------------
# ارزیابی چندکیسی
# -----------------------------
rows = []

for test_case in CASE_LIST:
    if case_data.get(test_case, {}).get("status") != "ok":
        rows.append({"test_case": test_case, "status": "skip_build_failed"})
        continue

    test_df = case_data[test_case]["model_df"]
    X_test = test_df[feature_cols_global]
    y_test = test_df["y"].astype(int)

    train_parts = []
    for train_case in CASE_LIST:
        if train_case == test_case:
            continue
        if case_data.get(train_case, {}).get("status") == "ok":
            train_parts.append(case_data[train_case]["model_df"])

    if len(train_parts) == 0:
        rows.append({"test_case": test_case, "status": "no_train_data"})
        continue

    train_df = pd.concat(train_parts, ignore_index=True)
    X_train = train_df[feature_cols_global]
    y_train = train_df["y"].astype(int)

    if y_train.nunique() < 2 or y_test.nunique() < 2:
        rows.append({"test_case": test_case, "status": "one_class_issue"})
        continue

    # --- آموزش مرحله اول و دوم ---
    model_stage1.fit(X_train, y_train)
    model_stage2.fit(X_train, y_train)

    prob1_train = model_stage1.predict_proba(X_train)[:, 1]
    prob1_test  = model_stage1.predict_proba(X_test)[:, 1]

    prob2_train = model_stage2.predict_proba(X_train)[:, 1]
    prob2_test  = model_stage2.predict_proba(X_test)[:, 1]

    # امتیاز جداسازی کلی (مرحله اول)
    auc1 = roc_auc_score(y_test, prob1_test)

    # --- تنظیم آستانه مرحله اول با هدف ۰٫۸۰ ---
    t1 = tune_threshold_by_recall(y_train, prob1_train, TARGET_RECALL_STAGE1)

    # --- تنظیم آستانه مرحله دوم (ملایم) با حفظ فراخوانی نهایی ---
    t2 = tune_stage2_threshold_soft(y_train, prob1_train, t1, prob2_train, TARGET_RECALL_FINAL)

    # --- سه حالت ارزیابی روی آزمون ---
    # (الف) فقط مرحله اول
    y_pred_s1 = (prob1_test >= t1).astype(int)
    m_s1 = eval_pred(y_test, y_pred_s1)

    # (ب) فقط مرحله دوم
    t2_alone = tune_threshold_by_recall(y_train, prob2_train, TARGET_RECALL_STAGE1)
    y_pred_s2 = (prob2_test >= t2_alone).astype(int)
    m_s2 = eval_pred(y_test, y_pred_s2)

    # (ج) دو مرحله‌ای
    y_pred_2stage = ((prob1_test >= t1) & (prob2_test >= t2)).astype(int)
    m_2s = eval_pred(y_test, y_pred_2stage)

    rows.append({
        "test_case": test_case,
        "status": "ok",
        "roc_auc_stage1": auc1,
        "t1": t1,
        "t2": t2,
        "t2_alone": t2_alone,
        **{f"s1_{k}": v for k, v in m_s1.items()},
        **{f"s2_{k}": v for k, v in m_s2.items()},
        **{f"two_{k}": v for k, v in m_2s.items()},
    })

two_stage_df = pd.DataFrame(rows)
two_stage_df
```

```python
import numpy as np
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import (
    confusion_matrix, roc_auc_score,
    precision_score, recall_score, f1_score,
    precision_recall_curve
)

# -----------------------------
# تنظیمات
# -----------------------------
BASE_PATH = r"E:\projects\Telemonitoring-Ai-Pipeline\data\uq-vitalsigns\uqvitalsignsdata"

THRESH_SPO2 = 98
HORIZON_SEC = 5 * 60
WIN_SEC = 60

CASE_LIST = [1,2,3,4,5]

TARGET_RECALL_STAGE1 = 0.80
TARGET_RECALL_FINAL  = 0.75


# -----------------------------
# خواندن داده
# -----------------------------
def load_case(case_id):
    cid = f"case{case_id:02d}"
    path = fr"{BASE_PATH}\{cid}\uq_vsd_{cid}_trenddata.csv"
    df = pd.read_csv(path, engine="python", on_bad_lines="skip")
    return df


# -----------------------------
# preprocess
# -----------------------------

def preprocess(df):
    d = df[["RelativeTimeMilliseconds","HR","SpO2"]].copy()

    d = d.apply(pd.to_numeric, errors="coerce")
    d = d.dropna(subset=["RelativeTimeMilliseconds"]).sort_values("RelativeTimeMilliseconds")

    # اگر داده خیلی کم بود → رد کن
    if len(d) < 200:
        return None

    diffs = d["RelativeTimeMilliseconds"].diff()
    diffs = diffs[diffs > 0]

    # ✅ اینجا fix اصلی
    if len(diffs) == 0:
        return None

    dt_ms = diffs.median()

    if pd.isna(dt_ms) or dt_ms <= 0:
        return None

    # resample
    if dt_ms < 500:
        d["t_bin"] = (d["RelativeTimeMilliseconds"] // 1000) * 1000
        d = d.groupby("t_bin", as_index=False).mean()

        diffs = d["RelativeTimeMilliseconds"].diff()
        diffs = diffs[diffs > 0]

        if len(diffs) == 0:
            return None

        dt_ms = diffs.median()

    return d, dt_ms


# -----------------------------
# feature + label
# -----------------------------
def build_dataset(d, dt_ms):
    h = int((HORIZON_SEC * 1000) / dt_ms)
    w = int((WIN_SEC * 1000) / dt_ms)

    future_min = d["SpO2"][::-1].rolling(h, min_periods=1).min()[::-1]
    d["y"] = (future_min < THRESH_SPO2).astype(int)

    d["HR_mean"] = d["HR"].rolling(w).mean()
    d["HR_std"]  = d["HR"].rolling(w).std()
    d["SpO2_mean"] = d["SpO2"].rolling(w).mean()
    d["SpO2_std"]  = d["SpO2"].rolling(w).std()

    feature_cols = ["HR_mean","HR_std","SpO2_mean","SpO2_std"]

    d = d.dropna()

    return d[feature_cols], d["y"]


# -----------------------------
# threshold selection (fast)
# -----------------------------
def tune_threshold_fast(y_true, prob, target_recall):
    p, r, t = precision_recall_curve(y_true, prob)

    df = pd.DataFrame({"t":np.r_[t, np.nan], "r":r, "p":p}).dropna()

    df = df[df["r"] >= target_recall]

    if len(df) == 0:
        return 0.5

    return df.sort_values("p", ascending=False).iloc[0]["t"]


# -----------------------------
# evaluation
# -----------------------------
def eval_pred(y, pred):
    tn, fp, fn, tp = confusion_matrix(y, pred).ravel()
    return {
        "precision": precision_score(y, pred, zero_division=0),
        "recall": recall_score(y, pred, zero_division=0),
        "f1": f1_score(y, pred, zero_division=0),
        "FP": fp,
        "FN": fn
    }


# -----------------------------
# models
# -----------------------------
model1 = GradientBoostingClassifier(random_state=42)

model2 = Pipeline([
    ("scaler", StandardScaler()),
    ("lr", LogisticRegression(max_iter=2000, class_weight="balanced"))
])


# -----------------------------
# run
# -----------------------------
results = []

for test_case in CASE_LIST:

    print(f"Running case {test_case}...")

    # split data
    train_X, train_y = [], []
    test_X, test_y = None, None

    for cid in CASE_LIST:
        df = load_case(cid)
        p = preprocess(df)

        if p is None:
            continue

        d, dt = p
        X, y = build_dataset(d, dt)

        if cid == test_case:
            test_X, test_y = X, y
        else:
            train_X.append(X)
            train_y.append(y)

    if test_X is None or len(train_X) == 0:
        continue

    X_train = pd.concat(train_X)
    y_train = pd.concat(train_y)

    if y_train.nunique() < 2 or test_y.nunique() < 2:
        continue

    # train models
    model1.fit(X_train, y_train)
    model2.fit(X_train, y_train)

    prob1_train = model1.predict_proba(X_train)[:,1]
    prob1_test  = model1.predict_proba(test_X)[:,1]

    prob2_train = model2.predict_proba(X_train)[:,1]
    prob2_test  = model2.predict_proba(test_X)[:,1]

    # thresholds
    t1 = tune_threshold_fast(y_train, prob1_train, TARGET_RECALL_STAGE1)
    t2 = tune_threshold_fast(y_train, prob2_train, TARGET_RECALL_FINAL)

    # predictions
    pred1 = (prob1_test >= t1).astype(int)
    pred2 = (prob2_test >= t2).astype(int)

    # two-stage
    pred_2stage = ((prob1_test >= t1) & (prob2_test >= t2)).astype(int)

    res = {
        "case": test_case,
        **{f"s1_{k}":v for k,v in eval_pred(test_y, pred1).items()},
        **{f"s2_{k}":v for k,v in eval_pred(test_y, pred2).items()},
        **{f"two_{k}":v for k,v in eval_pred(test_y, pred_2stage).items()},
    }

    results.append(res)

# dataframe
df_res = pd.DataFrame(results)
df_res
```

```python
import numpy as np
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import (
    confusion_matrix, roc_auc_score,
    precision_score, recall_score, f1_score,
    precision_recall_curve
)

# -----------------------------
# تنظیمات
# -----------------------------
BASE_PATH = r"E:\projects\Telemonitoring-Ai-Pipeline\data\uq-vitalsigns\uqvitalsignsdata"
THRESH_SPO2 = 98
HORIZON_SEC = 5 * 60
WIN_SEC = 60

CASE_LIST = [1, 2, 3, 4, 5]

TARGET_RECALL_STAGE1 = 0.80   # هدف فراخوانی مرحله اول
TARGET_RECALL_FINAL  = 0.75   # هدف فراخوانی نهایی (ملایم‌تر)


# -----------------------------
# خواندن فایل (نسخه پایدار)
# -----------------------------
def load_case_full(case_id: int):
    cid = f"case{case_id:02d}"
    path = fr"{BASE_PATH}\{cid}\uq_vsd_{cid}_trenddata.csv"
    df = pd.read_csv(path, sep=",", header=0, engine="python", on_bad_lines="skip", index_col=False)
    return df, path


# -----------------------------
# استخراج ۳ ستون + هم‌نمونه‌سازی اگر خیلی پرتراکم بود
# -----------------------------
def extract_and_resample(df):
    needed = ["RelativeTimeMilliseconds", "HR", "SpO2"]
    if not all(c in df.columns for c in needed):
        return None

    d = df[needed].copy()
    d["RelativeTimeMilliseconds"] = pd.to_numeric(d["RelativeTimeMilliseconds"], errors="coerce")
    d["HR"] = pd.to_numeric(d["HR"], errors="coerce")
    d["SpO2"] = pd.to_numeric(d["SpO2"], errors="coerce")
    d = d.dropna(subset=["RelativeTimeMilliseconds"]).sort_values("RelativeTimeMilliseconds")

    if len(d) < 200:
        return None

    diffs = d["RelativeTimeMilliseconds"].diff()
    diffs = diffs[diffs > 0]
    if len(diffs) == 0:
        return None

    dt_ms = diffs.median()
    if pd.isna(dt_ms) or dt_ms <= 0:
        return None

    # اگر داده خیلی پرتراکم بود -> 1 ثانیه‌ای
    if dt_ms < 500:
        d["t_bin"] = (d["RelativeTimeMilliseconds"] // 1000) * 1000
        d = d.groupby("t_bin", as_index=False).agg({
            "RelativeTimeMilliseconds": "first",
            "HR": "mean",
            "SpO2": "mean"
        }).sort_values("RelativeTimeMilliseconds")

        diffs = d["RelativeTimeMilliseconds"].diff()
        diffs = diffs[diffs > 0]
        if len(diffs) == 0:
            return None
        dt_ms = diffs.median()

    return d, float(dt_ms)


# -----------------------------
# ساخت برچسب + ویژگی‌های ۶۰ ثانیه‌ای (سبک قبلی و پایدار)
# -----------------------------
def build_dataset(d, dt_ms):
    h = int((HORIZON_SEC * 1000) / dt_ms)
    w = int((WIN_SEC * 1000) / dt_ms)

    future_min_spo2 = d["SpO2"][::-1].rolling(window=h, min_periods=1).min()[::-1]
    d = d.copy()
    d["y"] = (future_min_spo2 < THRESH_SPO2).astype(int)

    d["HR_mean"] = d["HR"].rolling(w, min_periods=w).mean()
    d["HR_std"]  = d["HR"].rolling(w, min_periods=w).std()
    d["HR_min"]  = d["HR"].rolling(w, min_periods=w).min()
    d["HR_max"]  = d["HR"].rolling(w, min_periods=w).max()

    d["SpO2_mean"] = d["SpO2"].rolling(w, min_periods=w).mean()
    d["SpO2_std"]  = d["SpO2"].rolling(w, min_periods=w).std()
    d["SpO2_min"]  = d["SpO2"].rolling(w, min_periods=w).min()
    d["SpO2_max"]  = d["SpO2"].rolling(w, min_periods=w).max()

    d["HR_slope"] = (d["HR"] - d["HR"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))
    d["SpO2_slope"] = (d["SpO2"] - d["SpO2"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))

    feature_cols = ["HR_mean","HR_std","HR_min","HR_max","HR_slope",
                    "SpO2_mean","SpO2_std","SpO2_min","SpO2_max","SpO2_slope"]

    model_df = d.dropna(subset=feature_cols + ["y"]).copy()
    return model_df, feature_cols


# -----------------------------
# تنظیم آستانه بر اساس هدف فراخوانی (مرحله ۱ و همچنین مرحله ۲ به صورت جداگانه)
# -----------------------------
def tune_threshold_by_recall(y_true, y_prob, target_recall):
    prec, rec, thr = precision_recall_curve(y_true, y_prob)
    pr_df = pd.DataFrame({"threshold": np.r_[thr, np.nan], "precision": prec, "recall": rec})
    c = pr_df.dropna(subset=["threshold"])
    c = c[c["recall"] >= target_recall]
    if len(c) == 0:
        return 0.5
    best = c.sort_values(["precision", "recall"], ascending=False).iloc[0]
    return float(best["threshold"])


# -----------------------------
# ارزیابی
# -----------------------------
def eval_pred(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    return {
        "precision_1": precision_score(y_true, y_pred, zero_division=0),
        "recall_1": recall_score(y_true, y_pred, zero_division=0),
        "f1_1": f1_score(y_true, y_pred, zero_division=0),
        "TN": tn, "FP": fp, "FN": fn, "TP": tp
    }


# -----------------------------
# تنظیم آستانه مرحله دوم (نسخه سریع: فقط ۵۰ آستانه نمونه‌برداری می‌کنیم)
# -----------------------------
def tune_stage2_threshold_soft_fast(y_train, prob1_train, t1, prob2_train, target_final_recall, max_checks=50):
    cand_mask = (prob1_train >= t1)
    if cand_mask.sum() == 0:
        return 0.5

    prec2, rec2, thr2 = precision_recall_curve(y_train[cand_mask], prob2_train[cand_mask])
    if len(thr2) == 0:
        return 0.5

    step = max(1, len(thr2)//max_checks)
    thr2_sample = thr2[::step]

    best_t2 = 0.5
    best_prec = -1.0

    for t2 in thr2_sample:
        y_pred_final = ((prob1_train >= t1) & (prob2_train >= t2)).astype(int)
        r = recall_score(y_train, y_pred_final, zero_division=0)
        p = precision_score(y_train, y_pred_final, zero_division=0)

        if r >= target_final_recall and p > best_prec:
            best_prec = p
            best_t2 = float(t2)

    return best_t2


# -----------------------------
# آماده‌سازی کیس‌ها (یک‌بار)
# -----------------------------
case_data = {}
feature_cols_global = None

for cid in CASE_LIST:
    df, path = load_case_full(cid)
    ext = extract_and_resample(df)
    if ext is None:
        case_data[cid] = {"status": "build_failed", "path": path}
        continue
    d, dt_ms = ext
    model_df, feature_cols = build_dataset(d, dt_ms)
    feature_cols_global = feature_cols
    case_data[cid] = {"status": "ok", "path": path, "model_df": model_df}


# -----------------------------
# دو مدل
# -----------------------------
stage1 = GradientBoostingClassifier(random_state=42)

stage2 = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("clf", LogisticRegression(max_iter=2000, class_weight="balanced", random_state=42))
])


# -----------------------------
# ارزیابی چندکیسی (هر بار یک کیس آزمون، بقیه آموزش)
# -----------------------------
rows = []

for test_case in CASE_LIST:
    if case_data.get(test_case, {}).get("status") != "ok":
        rows.append({"test_case": test_case, "status": "skip_build_failed"})
        continue

    test_df = case_data[test_case]["model_df"]
    X_test = test_df[feature_cols_global]
    y_test = test_df["y"].astype(int)

    train_parts = []
    for train_case in CASE_LIST:
        if train_case == test_case:
            continue
        if case_data.get(train_case, {}).get("status") == "ok":
            train_parts.append(case_data[train_case]["model_df"])

    if len(train_parts) == 0:
        rows.append({"test_case": test_case, "status": "no_train_data"})
        continue

    train_df = pd.concat(train_parts, ignore_index=True)
    X_train = train_df[feature_cols_global]
    y_train = train_df["y"].astype(int)

    if y_train.nunique() < 2 or y_test.nunique() < 2:
        rows.append({"test_case": test_case, "status": "one_class_issue"})
        continue

    # آموزش دو مدل
    stage1.fit(X_train, y_train)
    stage2.fit(X_train, y_train)

    prob1_train = stage1.predict_proba(X_train)[:, 1]
    prob1_test  = stage1.predict_proba(X_test)[:, 1]

    prob2_train = stage2.predict_proba(X_train)[:, 1]
    prob2_test  = stage2.predict_proba(X_test)[:, 1]

    auc1 = roc_auc_score(y_test, prob1_test)

    # آستانه مرحله اول (هدف فراخوانی)
    t1 = tune_threshold_by_recall(y_train, prob1_train, TARGET_RECALL_STAGE1)

    # آستانه مرحله دوم ملایم (نسخه سریع)
    t2 = tune_stage2_threshold_soft_fast(y_train, prob1_train, t1, prob2_train, TARGET_RECALL_FINAL, max_checks=50)

    # فقط مرحله اول
    y_pred_s1 = (prob1_test >= t1).astype(int)
    m_s1 = eval_pred(y_test, y_pred_s1)

    # فقط مرحله دوم (به صورت مستقل با هدف ۰٫۸۰)
    t2_alone = tune_threshold_by_recall(y_train, prob2_train, TARGET_RECALL_STAGE1)
    y_pred_s2 = (prob2_test >= t2_alone).astype(int)
    m_s2 = eval_pred(y_test, y_pred_s2)

    # دو مرحله‌ای (گزینه ۲: فیلتر ملایم)
    y_pred_two = ((prob1_test >= t1) & (prob2_test >= t2)).astype(int)
    m_two = eval_pred(y_test, y_pred_two)

    rows.append({
        "test_case": test_case,
        "status": "ok",
        "roc_auc_stage1": auc1,
        "t1": t1,
        "t2": t2,
        **{f"s1_{k}": v for k, v in m_s1.items()},
        **{f"s2_{k}": v for k, v in m_s2.items()},
        **{f"two_{k}": v for k, v in m_two.items()},
    })

two_stage_df = pd.DataFrame(rows)
two_stage_df
```

```python
import numpy as np
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import (
    confusion_matrix, roc_auc_score,
    precision_score, recall_score, f1_score,
    precision_recall_curve
)

# -----------------------------
# تنظیمات
# -----------------------------
BASE_PATH = r"E:\projects\Telemonitoring-Ai-Pipeline\data\uq-vitalsigns\uqvitalsignsdata"
THRESH_SPO2 = 98
HORIZON_SEC = 5 * 60
WIN_SEC = 60

CASE_LIST = [1, 2, 3, 4, 5]

TARGET_RECALL_STAGE1 = 0.80   # هدف فراخوانی مرحله اول
TARGET_RECALL_FINAL  = 0.80   # هدف فراخوانی نهایی (ملایم‌تر)


# -----------------------------
# خواندن فایل (نسخه پایدار)
# -----------------------------
def load_case_full(case_id: int):
    cid = f"case{case_id:02d}"
    path = fr"{BASE_PATH}\{cid}\uq_vsd_{cid}_trenddata.csv"
    df = pd.read_csv(path, sep=",", header=0, engine="python", on_bad_lines="skip", index_col=False)
    return df, path


# -----------------------------
# استخراج ۳ ستون + هم‌نمونه‌سازی اگر خیلی پرتراکم بود
# -----------------------------
def extract_and_resample(df):
    needed = ["RelativeTimeMilliseconds", "HR", "SpO2"]
    if not all(c in df.columns for c in needed):
        return None

    d = df[needed].copy()
    d["RelativeTimeMilliseconds"] = pd.to_numeric(d["RelativeTimeMilliseconds"], errors="coerce")
    d["HR"] = pd.to_numeric(d["HR"], errors="coerce")
    d["SpO2"] = pd.to_numeric(d["SpO2"], errors="coerce")
    d = d.dropna(subset=["RelativeTimeMilliseconds"]).sort_values("RelativeTimeMilliseconds")

    if len(d) < 200:
        return None

    diffs = d["RelativeTimeMilliseconds"].diff()
    diffs = diffs[diffs > 0]
    if len(diffs) == 0:
        return None

    dt_ms = diffs.median()
    if pd.isna(dt_ms) or dt_ms <= 0:
        return None

    # اگر داده خیلی پرتراکم بود -> 1 ثانیه‌ای
    if dt_ms < 500:
        d["t_bin"] = (d["RelativeTimeMilliseconds"] // 1000) * 1000
        d = d.groupby("t_bin", as_index=False).agg({
            "RelativeTimeMilliseconds": "first",
            "HR": "mean",
            "SpO2": "mean"
        }).sort_values("RelativeTimeMilliseconds")

        diffs = d["RelativeTimeMilliseconds"].diff()
        diffs = diffs[diffs > 0]
        if len(diffs) == 0:
            return None
        dt_ms = diffs.median()

    return d, float(dt_ms)


# -----------------------------
# ساخت برچسب + ویژگی‌های ۶۰ ثانیه‌ای (سبک قبلی و پایدار)
# -----------------------------
def build_dataset(d, dt_ms):
    h = int((HORIZON_SEC * 1000) / dt_ms)
    w = int((WIN_SEC * 1000) / dt_ms)

    future_min_spo2 = d["SpO2"][::-1].rolling(window=h, min_periods=1).min()[::-1]
    d = d.copy()
    d["y"] = (future_min_spo2 < THRESH_SPO2).astype(int)

    d["HR_mean"] = d["HR"].rolling(w, min_periods=w).mean()
    d["HR_std"]  = d["HR"].rolling(w, min_periods=w).std()
    d["HR_min"]  = d["HR"].rolling(w, min_periods=w).min()
    d["HR_max"]  = d["HR"].rolling(w, min_periods=w).max()

    d["SpO2_mean"] = d["SpO2"].rolling(w, min_periods=w).mean()
    d["SpO2_std"]  = d["SpO2"].rolling(w, min_periods=w).std()
    d["SpO2_min"]  = d["SpO2"].rolling(w, min_periods=w).min()
    d["SpO2_max"]  = d["SpO2"].rolling(w, min_periods=w).max()

    d["HR_slope"] = (d["HR"] - d["HR"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))
    d["SpO2_slope"] = (d["SpO2"] - d["SpO2"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))

    feature_cols = ["HR_mean","HR_std","HR_min","HR_max","HR_slope",
                    "SpO2_mean","SpO2_std","SpO2_min","SpO2_max","SpO2_slope"]

    model_df = d.dropna(subset=feature_cols + ["y"]).copy()
    return model_df, feature_cols


# -----------------------------
# تنظیم آستانه بر اساس هدف فراخوانی (مرحله ۱ و همچنین مرحله ۲ به صورت جداگانه)
# -----------------------------
def tune_threshold_by_recall(y_true, y_prob, target_recall):
    prec, rec, thr = precision_recall_curve(y_true, y_prob)
    pr_df = pd.DataFrame({"threshold": np.r_[thr, np.nan], "precision": prec, "recall": rec})
    c = pr_df.dropna(subset=["threshold"])
    c = c[c["recall"] >= target_recall]
    if len(c) == 0:
        return 0.5
    best = c.sort_values(["precision", "recall"], ascending=False).iloc[0]
    return float(best["threshold"])


# -----------------------------
# ارزیابی
# -----------------------------
def eval_pred(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    return {
        "precision_1": precision_score(y_true, y_pred, zero_division=0),
        "recall_1": recall_score(y_true, y_pred, zero_division=0),
        "f1_1": f1_score(y_true, y_pred, zero_division=0),
        "TN": tn, "FP": fp, "FN": fn, "TP": tp
    }


# -----------------------------
# تنظیم آستانه مرحله دوم (نسخه سریع: فقط ۵۰ آستانه نمونه‌برداری می‌کنیم)
# -----------------------------
def tune_stage2_threshold_soft_fast(y_train, prob1_train, t1, prob2_train, target_final_recall, max_checks=50):
    cand_mask = (prob1_train >= t1)
    if cand_mask.sum() == 0:
        return 0.5

    prec2, rec2, thr2 = precision_recall_curve(y_train[cand_mask], prob2_train[cand_mask])
    if len(thr2) == 0:
        return 0.5

    step = max(1, len(thr2)//max_checks)
    thr2_sample = thr2[::step]

    best_t2 = 0.5
    best_prec = -1.0

    for t2 in thr2_sample:
        y_pred_final = ((prob1_train >= t1) & (prob2_train >= t2)).astype(int)
        r = recall_score(y_train, y_pred_final, zero_division=0)
        p = precision_score(y_train, y_pred_final, zero_division=0)

        if r >= target_final_recall and p > best_prec:
            best_prec = p
            best_t2 = float(t2)

    return best_t2


# -----------------------------
# آماده‌سازی کیس‌ها (یک‌بار)
# -----------------------------
case_data = {}
feature_cols_global = None

for cid in CASE_LIST:
    df, path = load_case_full(cid)
    ext = extract_and_resample(df)
    if ext is None:
        case_data[cid] = {"status": "build_failed", "path": path}
        continue
    d, dt_ms = ext
    model_df, feature_cols = build_dataset(d, dt_ms)
    feature_cols_global = feature_cols
    case_data[cid] = {"status": "ok", "path": path, "model_df": model_df}


# -----------------------------
# دو مدل
# -----------------------------
stage1 = GradientBoostingClassifier(random_state=42)

stage2 = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("clf", LogisticRegression(max_iter=2000, class_weight="balanced", random_state=42))
])


# -----------------------------
# ارزیابی چندکیسی (هر بار یک کیس آزمون، بقیه آموزش)
# -----------------------------
rows = []

for test_case in CASE_LIST:
    if case_data.get(test_case, {}).get("status") != "ok":
        rows.append({"test_case": test_case, "status": "skip_build_failed"})
        continue

    test_df = case_data[test_case]["model_df"]
    X_test = test_df[feature_cols_global]
    y_test = test_df["y"].astype(int)

    train_parts = []
    for train_case in CASE_LIST:
        if train_case == test_case:
            continue
        if case_data.get(train_case, {}).get("status") == "ok":
            train_parts.append(case_data[train_case]["model_df"])

    if len(train_parts) == 0:
        rows.append({"test_case": test_case, "status": "no_train_data"})
        continue

    train_df = pd.concat(train_parts, ignore_index=True)
    X_train = train_df[feature_cols_global]
    y_train = train_df["y"].astype(int)

    if y_train.nunique() < 2 or y_test.nunique() < 2:
        rows.append({"test_case": test_case, "status": "one_class_issue"})
        continue

    # آموزش دو مدل
    stage1.fit(X_train, y_train)
    stage2.fit(X_train, y_train)

    prob1_train = stage1.predict_proba(X_train)[:, 1]
    prob1_test  = stage1.predict_proba(X_test)[:, 1]

    prob2_train = stage2.predict_proba(X_train)[:, 1]
    prob2_test  = stage2.predict_proba(X_test)[:, 1]

    auc1 = roc_auc_score(y_test, prob1_test)

    # آستانه مرحله اول (هدف فراخوانی)
    t1 = tune_threshold_by_recall(y_train, prob1_train, TARGET_RECALL_STAGE1)

    # آستانه مرحله دوم ملایم (نسخه سریع)
    t2 = tune_stage2_threshold_soft_fast(y_train, prob1_train, t1, prob2_train, TARGET_RECALL_FINAL, max_checks=50)

    # فقط مرحله اول
    y_pred_s1 = (prob1_test >= t1).astype(int)
    m_s1 = eval_pred(y_test, y_pred_s1)

    # فقط مرحله دوم (به صورت مستقل با هدف ۰٫۸۰)
    t2_alone = tune_threshold_by_recall(y_train, prob2_train, TARGET_RECALL_STAGE1)
    y_pred_s2 = (prob2_test >= t2_alone).astype(int)
    m_s2 = eval_pred(y_test, y_pred_s2)

    # دو مرحله‌ای (گزینه ۲: فیلتر ملایم)
    y_pred_two = ((prob1_test >= t1) & (prob2_test >= t2)).astype(int)
    m_two = eval_pred(y_test, y_pred_two)

    rows.append({
        "test_case": test_case,
        "status": "ok",
        "roc_auc_stage1": auc1,
        "t1": t1,
        "t2": t2,
        **{f"s1_{k}": v for k, v in m_s1.items()},
        **{f"s2_{k}": v for k, v in m_s2.items()},
        **{f"two_{k}": v for k, v in m_two.items()},
    })

two_stage_df = pd.DataFrame(rows)
two_stage_df
```

```python
import numpy as np
import pandas as pd
import warnings

from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import (
    confusion_matrix, precision_score, recall_score, f1_score,
    roc_auc_score, precision_recall_curve
)

# اگر پیام‌های هشدار خواندن فایل زیاد مزاحم هستند، این خط را فعال کن:
# warnings.filterwarnings("ignore", category=pd.errors.ParserWarning)

BASE_PATH = r"E:\projects\Telemonitoring-Ai-Pipeline\data\uq-vitalsigns\uqvitalsignsdata"
THRESH_SPO2 = 98
HORIZON_SEC = 5 * 60
WIN_SEC = 60

CASE_LIST = [1,2,3,4,5]

# چند هدفِ فراخوانی که می‌خواهیم امتحان کنیم
TARGET_RECALL_LIST = [0.65, 0.70, 0.75, 0.80]

def load_case_full(case_id: int):
    cid = f"case{case_id:02d}"
    path = fr"{BASE_PATH}\{cid}\uq_vsd_{cid}_trenddata.csv"
    df = pd.read_csv(path, sep=",", header=0, engine="python", on_bad_lines="skip", index_col=False)
    return df, path

def extract_and_resample(df):
    needed = ["RelativeTimeMilliseconds", "HR", "SpO2"]
    if not all(c in df.columns for c in needed):
        return None

    d = df[needed].copy()
    d["RelativeTimeMilliseconds"] = pd.to_numeric(d["RelativeTimeMilliseconds"], errors="coerce")
    d["HR"] = pd.to_numeric(d["HR"], errors="coerce")
    d["SpO2"] = pd.to_numeric(d["SpO2"], errors="coerce")
    d = d.dropna(subset=["RelativeTimeMilliseconds"]).sort_values("RelativeTimeMilliseconds")

    if len(d) < 200:
        return None

    diffs = d["RelativeTimeMilliseconds"].diff()
    diffs = diffs[diffs > 0]
    if len(diffs) == 0:
        return None

    dt_ms = diffs.median()
    if pd.isna(dt_ms) or dt_ms <= 0:
        return None

    # اگر داده خیلی پرتراکم بود -> ۱ ثانیه‌ای
    if dt_ms < 500:
        d["t_bin"] = (d["RelativeTimeMilliseconds"] // 1000) * 1000
        d = d.groupby("t_bin", as_index=False).agg({
            "RelativeTimeMilliseconds": "first",
            "HR": "mean",
            "SpO2": "mean"
        }).sort_values("RelativeTimeMilliseconds")

        diffs = d["RelativeTimeMilliseconds"].diff()
        diffs = diffs[diffs > 0]
        if len(diffs) == 0:
            return None
        dt_ms = diffs.median()

    return d, float(dt_ms)

def build_dataset(d, dt_ms):
    h = int((HORIZON_SEC * 1000) / dt_ms)
    w = int((WIN_SEC * 1000) / dt_ms)

    future_min_spo2 = d["SpO2"][::-1].rolling(window=h, min_periods=1).min()[::-1]
    d = d.copy()
    d["y"] = (future_min_spo2 < THRESH_SPO2).astype(int)

    # ----------- featureهای قبلی -----------
    d["HR_mean"] = d["HR"].rolling(w, min_periods=w).mean()
    d["HR_std"]  = d["HR"].rolling(w, min_periods=w).std()
    d["HR_min"]  = d["HR"].rolling(w, min_periods=w).min()
    d["HR_max"]  = d["HR"].rolling(w, min_periods=w).max()

    d["SpO2_mean"] = d["SpO2"].rolling(w, min_periods=w).mean()
    d["SpO2_std"]  = d["SpO2"].rolling(w, min_periods=w).std()
    d["SpO2_min"]  = d["SpO2"].rolling(w, min_periods=w).min()
    d["SpO2_max"]  = d["SpO2"].rolling(w, min_periods=w).max()

    d["HR_slope"] = (d["HR"] - d["HR"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))
    d["SpO2_slope"] = (d["SpO2"] - d["SpO2"].shift(w)) / (d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(w))

    # ----------- feature جدید (خیلی مهم) -----------

    # تغییر سریع SpO2
    d["SpO2_diff_5"] = d["SpO2"].diff(5)

    # مین کوتاه‌مدت
    short_w = max(3, int(w/3))
    d["SpO2_min_short"] = d["SpO2"].rolling(short_w, min_periods=1).min()

    # شیب کوتاه‌مدت
    d["SpO2_slope_short"] = (d["SpO2"] - d["SpO2"].shift(short_w)) / (
        d["RelativeTimeMilliseconds"] - d["RelativeTimeMilliseconds"].shift(short_w)
    )

    feature_cols = [
        "HR_mean","HR_std","HR_min","HR_max","HR_slope",
        "SpO2_mean","SpO2_std","SpO2_min","SpO2_max","SpO2_slope",
        "SpO2_diff_5","SpO2_min_short","SpO2_slope_short"
    ]

    model_df = d.dropna(subset=feature_cols + ["y"]).copy()
    return model_df, feature_cols
    
def tune_threshold_by_recall(y_true, y_prob, target_recall):
    p, r, t = precision_recall_curve(y_true, y_prob)
    pr = pd.DataFrame({"threshold": np.r_[t, np.nan], "precision": p, "recall": r}).dropna()
    pr = pr[pr["recall"] >= target_recall]
    if len(pr) == 0:
        return 0.5
    # بین گزینه‌هایی که فراخوانی کافی دارند، بیشترین دقت را انتخاب می‌کنیم
    return float(pr.sort_values(["precision", "recall"], ascending=False).iloc[0]["threshold"])

def eval_at_threshold(y_true, y_prob, thr):
    y_pred = (y_prob >= thr).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    return {
        "precision": precision_score(y_true, y_pred, zero_division=0),
        "recall": recall_score(y_true, y_pred, zero_division=0),
        "f1": f1_score(y_true, y_pred, zero_division=0),
        "FP": fp,
        "FN": fn,
        "TP": tp,
        "TN": tn
    }

# ---- آماده‌سازی همه کیس‌ها یک‌بار ----
case_data = {}
feature_cols_global = None

for cid in CASE_LIST:
    df, _ = load_case_full(cid)
    ext = extract_and_resample(df)
    if ext is None:
        case_data[cid] = {"status": "build_failed"}
        continue
    d, dt_ms = ext
    model_df, feature_cols = build_dataset(d, dt_ms)
    feature_cols_global = feature_cols
    case_data[cid] = {"status": "ok", "model_df": model_df}

# ---- ارزیابی چندکیسی برای هر هدف فراخوانی ----
rows = []
model = GradientBoostingClassifier(random_state=42)

for test_case in CASE_LIST:
    if case_data.get(test_case, {}).get("status") != "ok":
        continue

    test_df = case_data[test_case]["model_df"]
    X_test = test_df[feature_cols_global]
    y_test = test_df["y"].astype(int)

    train_parts = []
    for train_case in CASE_LIST:
        if train_case == test_case:
            continue
        if case_data.get(train_case, {}).get("status") == "ok":
            train_parts.append(case_data[train_case]["model_df"])

    if len(train_parts) == 0:
        continue

    train_df = pd.concat(train_parts, ignore_index=True)
    X_train = train_df[feature_cols_global]
    y_train = train_df["y"].astype(int)

    # اگر آموزش یا آزمون تک‌کلاسه باشد، این کیس را رد می‌کنیم
    if y_train.nunique() < 2 or y_test.nunique() < 2:
        continue

    model.fit(X_train, y_train)
    prob_train = model.predict_proba(X_train)[:, 1]
    prob_test  = model.predict_proba(X_test)[:, 1]

    auc = roc_auc_score(y_test, prob_test)

    for trg in TARGET_RECALL_LIST:
        thr = tune_threshold_by_recall(y_train, prob_train, trg)
        m = eval_at_threshold(y_test, prob_test, thr)
        rows.append({
            "test_case": test_case,
            "target_recall": trg,
            "threshold": thr,
            "roc_auc": auc,
            **m
        })

result_df = pd.DataFrame(rows)
result_df
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings

from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import (
    confusion_matrix, precision_score, recall_score, f1_score,
    roc_auc_score, precision_recall_curve
)

# اگر هشدارهای ParserWarning اذیت می‌کند، این خط را فعال کن:
# warnings.filterwarnings("ignore", category=pd.errors.ParserWarning)

# -----------------------------
# تنظیمات
# -----------------------------
BASE_PATH = r"E:\projects\Telemonitoring-Ai-Pipeline\data\uq-vitalsigns\uqvitalsignsdata"
THRESH_SPO2 = 98
HORIZON_SEC = 5 * 60
WIN_SEC = 60

CASE_LIST = [1,2,3,4,5]          # همین‌ها کافی است
CASE_ID = 3                      # کیس مورد تست (می‌توانی 1 یا 4 هم بگذاری)
TARGET_RECALL = 0.80             # همان تنظیم نهایی‌ات


# -----------------------------
# خواندن فایل کیس
# -----------------------------
def load_case_full(case_id: int):
    cid = f"case{case_id:02d}"
    path = fr"{BASE_PATH}\{cid}\uq_vsd_{cid}_trenddata.csv"
    df = pd.read_csv(path, sep=",", header=0, engine="python", on_bad_lines="skip", index_col=False)
    return df, path


# -----------------------------
# استخراج ۳ ستون + هم‌نمونه‌سازی
# -----------------------------
def extract_and_resample(df):
    needed = ["RelativeTimeMilliseconds", "HR", "SpO2"]
    if not all(c in df.columns for c in needed):
        return None

    d = df[needed].copy()
    d["RelativeTimeMilliseconds"] = pd.to_numeric(d["RelativeTimeMilliseconds"], errors="coerce")
    d["HR"] = pd.to_numeric(d["HR"], errors="coerce")
    d["SpO2"] = pd.to_numeric(d["SpO2"], errors="coerce")
    d = d.dropna(subset=["RelativeTimeMilliseconds"]).sort_values("RelativeTimeMilliseconds")

    if len(d) < 200:
        return None

    diffs = d["RelativeTimeMilliseconds"].diff()
    diffs = diffs[diffs > 0]
    if len(diffs) == 0:
        return None

    dt_ms = diffs.median()
    if pd.isna(dt_ms) or dt_ms <= 0:
        return None

    # اگر داده خیلی پرتراکم بود -> ۱ ثانیه‌ای
    if dt_ms < 500:
        d["t_bin"] = (d["RelativeTimeMilliseconds"] // 1000) * 1000
        d = d.groupby("t_bin", as_index=False).agg({
            "RelativeTimeMilliseconds": "first",
            "HR": "mean",
            "SpO2": "mean"
        }).sort_values("RelativeTimeMilliseconds")

        diffs = d["RelativeTimeMilliseconds"].diff()
        diffs = diffs[diffs > 0]
        if len(diffs) == 0:
            return None
        dt_ms = diffs.median()

    return d, float(dt_ms)


# -----------------------------
# تنظیم آستانه بر اساس هدف فراخوانی
# -----------------------------
def tune_threshold_by_recall(y_true, y_prob, target_recall):
    p, r, t = precision_recall_curve(y_true, y_prob)
    pr = pd.DataFrame({"threshold": np.r_[t, np.nan], "precision": p, "recall": r}).dropna()
    pr = pr[pr["recall"] >= target_recall]
    if len(pr) == 0:
        return 0.5
    return float(pr.sort_values(["precision","recall"], ascending=False).iloc[0]["threshold"])


# -----------------------------
# ارزیابی
# -----------------------------
def eval_at_threshold(y_true, y_prob, thr):
    y_pred = (y_prob >= thr).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    return {
        "precision": precision_score(y_true, y_pred, zero_division=0),
        "recall": recall_score(y_true, y_pred, zero_division=0),
        "f1": f1_score(y_true, y_pred, zero_division=0),
        "TN": tn, "FP": fp, "FN": fn, "TP": tp
    }


# -----------------------------
# ساخت دیتاست: نسخه قبل از بهبود (ویژگی‌های ساده)
# -----------------------------
def build_dataset_base(d, dt_ms):
    h = int((HORIZON_SEC * 1000) / dt_ms)
    w = int((WIN_SEC * 1000) / dt_ms)

    future_min_spo2 = d["SpO2"][::-1].rolling(window=h, min_periods=1).min()[::-1]
    dd = d.copy()
    dd["y"] = (future_min_spo2 < THRESH_SPO2).astype(int)

    # ویژگی‌های ساده
    dd["HR_mean"] = dd["HR"].rolling(w, min_periods=w).mean()
    dd["HR_std"]  = dd["HR"].rolling(w, min_periods=w).std()
    dd["SpO2_mean"] = dd["SpO2"].rolling(w, min_periods=w).mean()
    dd["SpO2_std"]  = dd["SpO2"].rolling(w, min_periods=w).std()

    feature_cols = ["HR_mean","HR_std","SpO2_mean","SpO2_std"]
    model_df = dd.dropna(subset=feature_cols + ["y"]).copy()
    return model_df, feature_cols


# -----------------------------
# ساخت دیتاست: نسخه بعد از بهبود (ویژگی‌های سریع)
# -----------------------------
def build_dataset_improved(d, dt_ms):
    h = int((HORIZON_SEC * 1000) / dt_ms)
    w = int((WIN_SEC * 1000) / dt_ms)

    future_min_spo2 = d["SpO2"][::-1].rolling(window=h, min_periods=1).min()[::-1]
    dd = d.copy()
    dd["y"] = (future_min_spo2 < THRESH_SPO2).astype(int)

    # ویژگی‌های قبلی + چند ویژگی برای افت سریع
    dd["HR_mean"] = dd["HR"].rolling(w, min_periods=w).mean()
    dd["HR_std"]  = dd["HR"].rolling(w, min_periods=w).std()
    dd["SpO2_mean"] = dd["SpO2"].rolling(w, min_periods=w).mean()
    dd["SpO2_std"]  = dd["SpO2"].rolling(w, min_periods=w).std()

    # ویژگی‌های جدید برای افت سریع
    dd["SpO2_diff_5"] = dd["SpO2"].diff(5)  # تغییر سریع در ۵ نمونه

    short_w = max(3, int(w/3))
    dd["SpO2_min_short"] = dd["SpO2"].rolling(short_w, min_periods=short_w).min()

    dd["SpO2_slope_short"] = (dd["SpO2"] - dd["SpO2"].shift(short_w)) / (
        dd["RelativeTimeMilliseconds"] - dd["RelativeTimeMilliseconds"].shift(short_w)
    )

    feature_cols = ["HR_mean","HR_std","SpO2_mean","SpO2_std",
                    "SpO2_diff_5","SpO2_min_short","SpO2_slope_short"]
    model_df = dd.dropna(subset=feature_cols + ["y"]).copy()
    return model_df, feature_cols


# -----------------------------
# آموزش و پیش‌بینی (کمکی)
# -----------------------------
def fit_predict_for_case(case_id, build_fn):
    # داده کیس آزمون
    df_test, _ = load_case_full(case_id)
    ext_test = extract_and_resample(df_test)
    if ext_test is None:
        return None

    d_test, dt_test = ext_test
    test_df, feature_cols = build_fn(d_test, dt_test)
    X_test = test_df[feature_cols]
    y_test = test_df["y"].astype(int)

    # داده آموزش از بقیه کیس‌ها
    train_parts = []
    for cid in CASE_LIST:
        if cid == case_id:
            continue
        df, _ = load_case_full(cid)
        ext = extract_and_resample(df)
        if ext is None:
            continue
        d, dt = ext
        tr_df, _ = build_fn(d, dt)
        train_parts.append(tr_df)

    if len(train_parts) == 0:
        return None

    train_df = pd.concat(train_parts, ignore_index=True)
    X_train = train_df[feature_cols]
    y_train = train_df["y"].astype(int)

    # اگر تک‌کلاسه شد، تست معنی ندارد
    if y_train.nunique() < 2 or y_test.nunique() < 2:
        return None

    model = GradientBoostingClassifier(random_state=42)
    model.fit(X_train, y_train)

    prob_train = model.predict_proba(X_train)[:, 1]
    prob_test  = model.predict_proba(X_test)[:, 1]

    thr = tune_threshold_by_recall(y_train, prob_train, TARGET_RECALL)
    metrics = eval_at_threshold(y_test, prob_test, thr)

    pred = (prob_test >= thr).astype(int)

    out = {
        "test_df": test_df,
        "feature_cols": feature_cols,
        "y_test": y_test.to_numpy(),
        "prob_test": prob_test,
        "pred": pred,
        "threshold": thr,
        "metrics": metrics,
    }
    return out


# -----------------------------
# اجرای تست قبل/بعد
# -----------------------------
base_out = fit_predict_for_case(CASE_ID, build_dataset_base)
imp_out  = fit_predict_for_case(CASE_ID, build_dataset_improved)

if base_out is None or imp_out is None:
    print("این کیس برای مقایسه آماده نشد (ممکن است تک‌کلاسه یا داده نامعتبر باشد).")
else:
    print("=== قبل از بهبود (ویژگی‌های ساده) ===")
    print("آستانه:", base_out["threshold"])
    print(base_out["metrics"])

    print("\n=== بعد از بهبود (ویژگی‌های سریع) ===")
    print("آستانه:", imp_out["threshold"])
    print(imp_out["metrics"])

    # -----------------------------
    # مقایسه عددی
    # -----------------------------
    pred_before = base_out["pred"]
    pred_after  = imp_out["pred"]
    y_true      = imp_out["y_test"]  # هر دو برچسب یکسان

    added_alerts = np.sum((pred_before == 0) & (pred_after == 1))
    removed_alerts = np.sum((pred_before == 1) & (pred_after == 0))

    # افت‌های واقعی که قبلاً جا افتاده بودند ولی الان گرفته شدند:
    recovered = np.sum((y_true == 1) & (pred_before == 0) & (pred_after == 1))
    # افت‌های واقعی که قبلاً گرفته می‌شدند ولی الان جا افتادند:
    lost = np.sum((y_true == 1) & (pred_before == 1) & (pred_after == 0))

    print("\n=== مقایسه تغییرات ===")
    print("هشدارهای اضافه شده (۰→۱):", added_alerts)
    print("هشدارهای حذف شده (۱→۰):", removed_alerts)
    print("افت‌های بازیابی‌شده (قبلاً جاافتاده، الان گرفته):", recovered)
    print("افت‌های از دست‌رفته (قبلاً گرفته، الان جاافتاده):", lost)

    # -----------------------------
    # نمودار ۱: احتمال‌ها قبل و بعد + آستانه‌ها
    # -----------------------------
    plt.figure(figsize=(15,5))
    plt.plot(base_out["prob_test"], label="احتمال قبل از بهبود", alpha=0.7)
    plt.plot(imp_out["prob_test"], label="احتمال بعد از بهبود", alpha=0.7)
    plt.axhline(base_out["threshold"], linestyle="--", label="آستانه قبل")
    plt.axhline(imp_out["threshold"], linestyle="--", label="آستانه بعد")
    plt.title(f"کیس {CASE_ID} — مقایسه احتمال‌های مدل قبل و بعد از بهبود")
    plt.xlabel("اندیس زمان")
    plt.ylabel("احتمال افت")
    plt.legend()
    plt.show()

    # -----------------------------
    # نمودار ۲: SpO2 + هشدارها قبل و بعد (روی یک نمودار)
    # -----------------------------
    spo2 = imp_out["test_df"]["SpO2"].to_numpy()  # سری واقعی اکسیژن
    t = np.arange(len(spo2))

    plt.figure(figsize=(15,6))
    plt.plot(t, spo2, label="SpO2 واقعی", alpha=0.7)
    plt.axhline(THRESH_SPO2, linestyle="--", label="آستانه ۹۸")

    # افت واقعی
    true_idx = np.where(y_true == 1)[0]
    plt.scatter(true_idx, spo2[true_idx], color="red", s=10, label="افت واقعی")

    # هشدارهای قبل
    idx_b = np.where(pred_before == 1)[0]
    plt.scatter(idx_b, spo2[idx_b], color="orange", s=12, label="هشدار قبل (قدیمی)")

    # هشدارهای بعد
    idx_a = np.where(pred_after == 1)[0]
    plt.scatter(idx_a, spo2[idx_a], color="green", s=12, label="هشدار بعد (بهبود یافته)")

    plt.title(f"کیس {CASE_ID} — مقایسه هشدارهای قبل و بعد روی سیگنال واقعی")
    plt.xlabel("اندیس زمان")
    plt.ylabel("SpO2")
    plt.legend()
    plt.show()
```
